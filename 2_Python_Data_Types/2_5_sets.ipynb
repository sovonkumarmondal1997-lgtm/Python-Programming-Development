1. Sets Functions

```python
# len()
my_set = {1, 2, 3, 4, 5}
print("Length of set:", len(my_set))
```

```python
# sum()
my_set = {1, 2, 3, 4, 5}
print("sum of set:", sum(my_set))
```

```python
# max() and min()
my_set = {1, 2, 3, 4, 5}
print(max(my_set))
print(min(my_set))
```

```python
# sorted()
my_set = {5, 3, 1, 4, 2}
print(sorted(my_set))
print(sorted(my_set, reverse=True))
```

```python
# union() method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.union(set2))
print(set2.union(set1))
```

```python
# update() method in sets: perform union and update the set and it's a permanent change to the set
set1 = {1, 2, 3}
set2 = {4, 5, 6}

set1.update(set2)
print("set1 after update:", set1)
print("current value of set2:", set2)

set2.update(set1)
print("current value of set1:", set1)
print("set2 after update:", set2)
```

```python
# intersection() method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.intersection(set2))
print(set2.intersection(set1))
```

```python
# intersection_update() method in sets: perform intersection and update the set and it's a permanent change to the set
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set1.intersection_update(set2)
print("set1 after intersection_update:", set1)
print("current value of set2:", set2)

set2.intersection_update(set1)
print("current value of set1:", set1)
print("set2 after intersection_update:", set2)
```

```python
# difference() method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
# difference_update() method in sets: perform difference and update the set and it's a permanent change to the set
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.difference_update(set2)
print("set1 after difference_update:", set1)
print("current value of set2:", set2)

set2.difference_update(set1)
print("current value of set1:", set1)
print("set2 after difference_update:", set2)
```

```python
# symmetric_difference() method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.symmetric_difference(set2))
print(set2.symmetric_difference(set1))
```

```python
# symmetric_difference_update() method in sets: perform symmetric difference and update the set and it's a permanent change to the set
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.symmetric_difference_update(set2)
print("set1 after symmetric_difference_update:", set1)
print("current value of set2:", set2)

set2.symmetric_difference_update(set1)
print("current value of set1:", set1)
print("set2 after symmetric_difference_update:", set2)
```
