```python
# what is function?:
# -> function is a block of code which only runs when it is called. you can pass data, known as parameters,
# -> into a function. a function can return data as a result.

# different types of functions:
# 1. built-in functions 
# 2. user-defined functions

# principal of functions:
# 1. DRY (Don't Repeat Yourself)
# 2. modularity (divide and conquer)
# 3. reusability (write once, use many times)
# 4. abstraction (hide complexity)
# 5. maintainability (easy to update and fix bugs)

# components of a function:
# 1. function keyword: the keyword used to define a function (def in Python).
# 2. function name: a unique identifier for the function.
# 3. parameters or arguments (input): variables that are passed to the function.
# 4. docstring: a string that describes the purpose and usage of the function (optional but recommended).
# 5. function body: the block of code that performs the task.
# 6. return statement: the value that the function returns after execution.
# 7. function call: the statement that invokes the function to execute its code.
```

```python
# let's create a simple function that takes two numbers as input and returns their sum with docstring:
def add_numbers(a, b):
    """
    This function takes two numbers as input and returns their sum.
    
    Parameters:
    a (int or float): The first number.
    b (int or float): The second number.
    
    Returns:
    int or float: The sum of the two numbers.
    """
    return a + b

# let's call the function and print the result:
result = add_numbers(5, 3)
print(result)
```

```python
def sum_numbers(a,b):
    """ 
    This function takes two numbers as input and returns their sum. 

    Parameters:
    a (int or float): The first number.
    b (int or float): The second number.
    
    Returns:
    int or float: The sum of the two numbers.
    """
    sum = a + b
    return sum

result = sum_numbers(5, 5)
print(result)
```

```python
# let's create a simple function for calculating odd or even number using docstring:
def odd_or_even(num):
    """
    This function takes a number as input and returns whether it is odd or even.
    
    Parameters:
    num (int): The number to check.
    
    Returns:
    str: A string indicating whether the number is odd or even.
    """
    if type(num) == int:
        if num % 2 == 0:
            return "Even"
        else:
            return "Odd"
    else:
        return "Invalid input. Please enter an integer."
    
  
# let's call the function and print the result:
number = 10
result = odd_or_even(number)
print(f"{number} is {result}.")
```

```python
result = odd_or_even(7)
print(result)
```

```python
# call the odd_or_even function with a loop to check numbers from 1 to 10:
for i in range(1, 11):
    result = odd_or_even(i)
    print(f"{i} is {result}.")
```

```python
"""
1. Parameter in function:
=> parameters are variables that are defined in the function definition and are used to receive input values,
   when the function is called. they act as placeholders for the values that will be passed to the function.
"""
# exampl of a parameter in a function:
def greet(name):
    """
    This function takes a name as input and returns a greeting message.
    
    Parameters:
    name (str): The name of the person to greet.
    
    Returns:
    str: A greeting message.
    """
    return f"Hello, {name}!"

# let's call the function and print the result:
greeting = greet("Alice")
print(greeting)
```

```python
"""
2. Argument in function:
=> arguments are the actual values that are passed to the function when it is called.
   they correspond to the parameters defined in the function.
"""
# example of an argument in a function:
greeting = greet("Alice")
print(greeting)
```

```python
# Types of arguments in a function:
# 1. Positional arguments: these are the most common type of arguments where the values are passed in the same order,
#    as the parameters defined in the function.

# 2. Keyword arguments: these allow you to specify the values for the parameters by using the parameter names, 
#    regardless of their order in the function definition.

# 3. Default arguments: these are parameters that have a default value assigned to them.
#    if no value is provided for that parameter when the function is called.

# 4. Variable-length arguments: these allow you to pass a variable number of arguments to a function.
#    there are two types of variable-length arguments: *args for non-keyword arguments and **kwargs for keyword arguments
```

```python
# 1. Example of positional arguments:
def multiply(x, y):
    """
    This function takes two numbers as input and returns their product.
    
    Parameters:
    x (int or float): The first number.
    y (int or float): The second number.
    
    Returns:
    int or float: The product of the two numbers.
    """
    return x * y

# Calling the function with positional arguments
result = multiply(5, 3)
print(result)
```

```python
# 2. Example of keyword arguments:
def divide(x, y):
    """
    This function takes two numbers as input and returns their quotient.
    
    Parameters:
    x (int or float): The first number.
    y (int or float): The second number.
    
    Returns:
    int or float: The quotient of the two numbers.
    """
    return x / y

# Calling the function with keyword arguments
result = divide(y=5, x=10)
print(result)
```

```python
# 3. Example of default arguments:
def power(base, exponent=2):
    """
    This function takes a base number and an optional exponent and returns the result of base raised to the power of exponent.
    
    Parameters:
    base (int or float): The base number.
    exponent (int or float, optional): The exponent to which the base is raised. Default is 2.
    
    Returns:
    int or float: The result of base raised to the power of exponent.
    """
    return base ** exponent

# Calling the function with the default exponent
result1 = power(5)
print(result1)

# Calling the function with a specified exponent
result2 = power(5, 3)
print(result2)
```

```python
# 4. Example of variable-length arguments:
def sum_all(*args):
    """
    This function takes a variable number of arguments and returns their sum.

    Parameters:
    *args: A variable number of arguments to be summed.

    Returns:
    int: The sum of all the arguments.
    """
    return sum(args)

# Calling the function with a variable number of arguments
result = sum_all(1, 2, 3, 4, 5)
print(result)
```

```python
# *args: A tuple of all positional arguments or variable number of arguments passed to the function
# example of *args with a function that calculate the sum of all the numbers passed as arguments:
def sum_all(*args):
    """
    This function takes a variable number of arguments and returns their sum.

    Parameters:
    *args: A variable number of arguments to be summed.

    Returns:
    int: The sum of all the arguments.
    """
    return sum(args)

# Calling the function with a variable number of arguments
result = sum_all(1, 2, 3, 4, 5)
print(result)
```

```python
# example of *args with a function that calculate the multiplication of all the numbers passed as arguments:
def multiply_all(*args):
    """
    This function takes a variable number of arguments and returns their product.

    Parameters:
    *args: A variable number of arguments to be multiplied.

    Returns:
    int: The product of all the arguments.
    """
    result = 1
    for num in args:
        result = result * num
    return result

# Calling the function with a variable number of arguments
result = multiply_all(1, 2, 3, 4, 5)
print(result)

# Calling the function with a different set of arguments
result1 = multiply_all(5,6,7,8,9,10)
print(result1)
```

```python
# **kwargs: A dictionary of all keyword arguments passed to the function
def print_info(**kwargs):
    """
    This function takes a variable number of keyword arguments and prints the key-value pairs.

    Parameters:
    **kwargs: A variable number of keyword arguments to be printed.
    """
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Calling the function with keyword arguments
print_info(name="Alice", age=30, city="New York")
```

```python
# Difference between return and print in a function:

# 1. Return: the return statement is used to exit a function and return a value to the caller.
# it allows you to capture the output of a function and use it elsewhere in your code.

# 2. Print: the print function is used to display output to the console.
# it does not return a value and cannot be captured or used in other parts of the code.

# example of return vs print in a function:
def add_and_print(a, b):
    print(a + b)
def add_and_return(a, b):
    return a + b

# Calling the functions
add_and_print(5, 3)  # This will print the result but return None

result = add_and_return(5, 3)  # This will return the result and store it in the variable 'result'
print(result)  # This will print the result returned by the add_and_return function
```
