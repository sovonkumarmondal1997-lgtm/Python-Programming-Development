```python
"""
Problem 1: Count Even Numbers in a List

Write a program that takes a list of integers from the user and prints how many even numbers are present in the list.

Example 1:

Input: [2, 5, 8, 11, 14, 17]
Output: Even numbers = 3

Example 2:

Input: [1, 3, 5, 7]
Output: Even numbers = 0

Requirements:
- Use a for loop.
- Do not use filter().
- Do not use list comprehension.
- Do not use any built-in function specifically for counting even numbers.
- Solve according to your current learning level.
"""
```

```python
L = [2, 5, 8, 11, 14, 17]
count = 0
for i in L:
    if i % 2 == 0:
        count += 1
print(f"Even numbers = {count}")
```

```python
"""
Problem 2: Find the Largest Element in a List

Write a program that finds the largest number in a list without using max()

Input: [12, 45, 7, 89, 23]
Output: Largest element = 89

"""
```

```python
input_list =[12, 45, 7, 89, 23]
largest_element = input_list[0]
for i in input_list:
    if i > largest_element:
        largest_element = i
print(f"Largest element = {largest_element}")
```

```python
"""
Problem 3: Count Positive, Negative, and Zero

Write a program that takes a list of integers and counts:

- Positive numbers
- Negative numbers
- Zero(s)

Print all three counts.

Example:

Input: [10, -5, 0, 8, -2, 0, 15, 0, 3, 6, -9, 2, -11, -4, 0, 6, 5, 14, -8]
Output:

Positive numbers = 3
Negative numbers = 2
Zeroes = 2
"""
```

```python
List_1 = [10, -5, 0, 8, -2, 0, 15, 0, 3, 6, -9, 2, -11, -4, 0, 6, 5, 14, -8]

positive_count = 0
negative_count = 0
zero_count = 0

for i in List_1:
    if i > 0:
        positive_count += 1
    elif i < 0:
        negative_count += 1
    else:
        zero_count += 1

print(f"Positive numbers = {positive_count}")
print(f"Negative numbers = {negative_count}")
print(f"Zeroes = {zero_count}")
```

```python
"""
Problem 5: Remove All Occurrences of a Given Number. Write a program that removes all occurrences of a given number from a list.

Example:

Input: numbers = [10, 20, 30, 20, 40, 20, 50]
target = 20

Output: [10, 30, 40, 50]

Requirements:
- Create a new list.
- Use only one for loop.
Do not use:
- remove()
- pop()
- filter()
- list comprehension
"""
```

```python
numbers = [10, 20, 30, 20, 40, 20, 50]
target = 20
new_list = []
for i in numbers:
    if i != target:
        new_list.append(i)
print(new_list)
```

```python
"""
Problem 6:
Find the Second Largest Element in a List. Write a program to find the second largest unique element in a list.

Example:

Input: numbers = [10, 20, 30, 40, 50, 60, 80, 40, 70, 90]
Output: Second largest element = 40

Requirements:
Do not use:
- sort()
- sorted()
- max()
- set()
- Traverse the list only once.
- Use only the concepts you've learned so far.
- Handle duplicate values correctly.
"""
```

```python
numbers = [10, 20, 30, 40, 50, 60, 80, 40, 70, 90]

largest = numbers[0]
second_largest = None

for num in numbers:

    if num > largest:
        second_largest = largest
        largest = num

    elif num != largest:
        if second_largest is None or num > second_largest:
            second_largest = num

if second_largest is None:
    print("Second largest element does not exist.")
else:
    print(f"Second largest element = {second_largest}")
```

```python
"""
Problem 7: Remove Duplicate Elements Without Using set()
Write a program that removes all duplicate elements from a list while preserving the order of their first occurrence.

Example:

Input: numbers = [10, 20, 10, 30, 20, 40, 30]
Output: [10, 20, 30, 40]

Requirements
Do not use:
set()
dict
list comprehension
Create a new list.
Preserve the order of the first occurrence.
"""
```

```python
"""
Problem 8: Left Rotate a List by One Position. Write a program that rotates a list to the left by one position.

Example:

Input: numbers = [10, 20, 30, 40, 50]
Output: [20, 30, 40, 50, 10]

Explanation:

The first element (10) moves to the end.
Every other element shifts one position to the left.

Example 2:

Input: numbers = [5]
Output: [5]

Requirements
Do not use slicing.
Do not use pop().
Do not use collections.deque.
You may create a new list.
"""
```

```python
numbers = [10, 20, 30, 40, 50]

new_list = []

for i in range(1, len(numbers)):
    new_list.append(numbers[i])

new_list.append(numbers[0])

print(new_list)
```

```python
"""
Problem 9: Find the Most Frequent Element in a List
Write a program to find the element that appears the most times in a list.
If multiple elements have the same highest frequency, print the one that appears first in the original list.

Example 1:

Input: numbers = [10, 20, 10, 30, 20, 10, 40]
Output: Most frequent element = 10
Frequency = 3

Example 2:

Input: numbers = [5, 2, 5, 2]
Output: Most frequent element = 5
Frequency = 2

Explanation: Both 5 and 2 appear 2 times, but 5 appears first in the list

Requirements
❌ Do not use:
- count()
- dict
- set
- collections.Counter
✅ Use nested loops.
✅ Use state tracking.
✅ Solve according to your current learning level
"""
```

```python
numbers = [10, 20, 10, 30, 20, 10, 40]
count = 0
for i in numbers:
    current_count = 0
    for j in numbers:
        if i == j:
            current_count += 1
    if current_count > count:
        count = current_count
        most_frequent = i
print(f"Most frequent element = {most_frequent}")
print(f"Frequency = {count}")
```

```python
"""
Problem 10: Merge Two Sorted Lists into One Sorted List.
Write a program that merges two already sorted lists into a single sorted list.

Example 1:

Input:

list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]

Output: [1, 2, 3, 4, 5, 6, 7, 8]

Example 2:

Input:

list1 = [5, 10, 15]
list2 = [2, 8, 20]

Output: [2, 5, 8, 10, 15, 20]

Example 3:

Input:

list1 = [1, 2, 3]
list2 = []

Output: [1, 2, 3]

Requirements:
❌ Do not use:
sort()
sorted()
extend()
list comprehension
✅ Create a new list.
✅ Preserve the sorted order.
"""
```

```python
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]

new_list = []

i = 0
j = 0

while i < len(list1) and j < len(list2):

    if list1[i] < list2[j]:
        new_list.append(list1[i])
        i += 1
    else:
        new_list.append(list2[j])
        j += 1

while i < len(list1):
    new_list.append(list1[i])
    i += 1

while j < len(list2):
    new_list.append(list2[j])
    j += 1

print(new_list)
```
