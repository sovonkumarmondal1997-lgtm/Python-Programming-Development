```python
"""
Problem 1 : Remove Duplicate Fruits While Preserving Order

You are given a tuple of fruits.

fruits = (
    "apple",
    "banana",
    "apple",
    "orange",
    "banana",
    "grape",
    "apple"
)
Create a new tuple that contains only the first occurrence of each fruit while preserving the original order.

Rules:
❌ Do not use set() to directly remove duplicates.
❌ Do not use dictionary keys to solve the problem.
✅ You may use a list as temporary storage if needed.
✅ Use loops.
✅ Finally convert the result back to a tuple
"""
```

```python
fruits = (
    "apple",
    "banana",
    "apple",
    "orange",
    "banana",
    "grape",
    "apple"
)

result = []

for fruit in fruits:
    if fruit not in result:
        result.append(fruit)

print(tuple(result))
```

```python
"""
Problem 2 : Find the Most Frequent Color
You are given a tuple of colors.

colors = (
    "red",
    "blue",
    "red",
    "green",
    "blue",
    "red",
    "yellow"
)

Find the color that appears the maximum number of times.

Expected Output: 
Most frequent color: red
Count: 3

Rules:
Do not use a dictionary.
Do not use count().
Do not convert the tuple to a set.
Use loops only.
If there is a tie, print the color that appears first in the tuple.

"""
```

```python
colors = (
    "red",
    "blue",
    "red",
    "green",
    "blue",
    "red",
    "yellow"
)

max_count = 0
most_frequent_color = ""

for color in colors:
    current_count = 0

    for item in colors:
        if color == item:
            current_count += 1

    if current_count > max_count:
        max_count = current_count
        most_frequent_color = color

print("Most frequent color:", most_frequent_color)
print("Count:", max_count)
```

```python
"""
Problem 3 : Find the Common Subjects

Two students have selected different subjects.

student1 = (
    "Math",
    "Physics",
    "Chemistry",
    "English",
    "Biology"
)

student2 = (
    "History",
    "Physics",
    "Biology",
    "Geography",
    "Math"
)

Print all the common subjects between the two tuples.

The order should follow the order of student1.

Expected Output:
Math
Physics
Biology

Rules:
Do not convert the tuples into sets.
Do not use intersection (&).
Do not use dictionaries.
Use loops only.
Print each common subject only once.
"""
```

```python
student1 = (
    "Math",
    "Physics",
    "Chemistry",
    "English",
    "Biology"
)

student2 = (
    "History",
    "Physics",
    "Biology",
    "Geography",
    "Math"
)

for subject in student1:
    if subject in student2:
        print(subject)
```

```python
"""
Problem 4: Count Unique Numbers in a Tuple

You are given a tuple of integers.

numbers = (
    10,
    20,
    10,
    30,
    40,
    20,
    50,
    30,
    60
)

Count how many unique numbers are present in the tuple.

Expected Output: 
Unique numbers: 6

Rules
Do not use set().
Do not use a dictionary.
Do not use count().
Use loops only.
Use a list if needed.
"""
```

```python
numbers = (
    10,
    20,
    10,
    30,
    40,
    20,
    50,
    30,
    60
)

unique_numbers = []

for number in numbers:
    if number not in unique_numbers:
        unique_numbers.append(number)

print("Unique numbers:", len(unique_numbers))
```

```python
"""
Problem 6:Employee Department Count

You are given a tuple containing employee names and their departments.

employees = (
    ("Rahul", "IT"),
    ("Anjali", "HR"),
    ("Rohit", "IT"),
    ("Sneha", "Finance"),
    ("Amit", "HR"),
    ("Priya", "IT"),
    ("Karan", "Finance")
)


Create a dictionary where:

The key is the department name.
The value is the number of employees in that department.

Expected Output:
{
    "IT": 3,
    "HR": 2,
    "Finance": 2
}

Rules:
Do not use collections.Counter.
Do not use setdefault().
Do not use dictionary comprehension.
Do not use list comprehension.
Use loops only.

"""
```

```python
"""
Problem 7 : Find Students Enrolled in Both Courses

Two courses have the following students:

python_students = (
    "Rahul",
    "Anjali",
    "Rohit",
    "Sneha",
    "Amit",
    "Priya"
)

sql_students = (
    "Priya",
    "Rohit",
    "Rahul",
    "Karan",
    "Anjali",
    "Rahul"
)

Create a list containing the names of students who are enrolled in both courses.

Expected Output:
['Rahul', 'Anjali', 'Rohit', 'Priya']

The order should follow python_students.

Rules:
Do not use set().
Do not use intersection (&).
Do not use a dictionary.
Do not use list comprehension.
Use loops only.
Each student should appear only once in the output.
"""
```

```python
python_students = (
    "Rahul",
    "Anjali",
    "Rohit",
    "Sneha",
    "Amit",
    "Priya"
)

sql_students = (
    "Priya",
    "Rohit",
    "Rahul",
    "Karan",
    "Anjali",
    "Rahul"
)

common_students = []

for student in python_students:
    if student in sql_students and student not in common_students:
        common_students.append(student)

print(common_students)
```

```python
"""
Problem 8 : Find the Department with the Maximum Employees

You are given a dictionary where:

Key = Department name
Value = Number of employees

departments = {
    "IT": 12,
    "HR": 5,
    "Finance": 8,
    "Sales": 15,
    "Marketing": 9
}

Find the department that has the maximum number of employees.

Expected Output:
Department: Sales
Employees: 15

Rules:
Do not use max().
Do not sort the dictionary.
Do not convert the dictionary into a list.
Do not use list comprehension.
Use loops only.
If there is a tie, print the department that appears first in the dictionary.

"""
```

```python
departments = {
    "IT": 12,
    "HR": 5,
    "Finance": 8,
    "Sales": 15,
    "Marketing": 9
}

max_employees = 0
department_with_max_employees = ""

for department, employees in departments.items():
    if employees > max_employees:
        max_employees = employees
        department_with_max_employees = department

print(f"Department: {department_with_max_employees}")
```

```python
"""
Problem 9 : Calculate Total Sales for Each Product

You are given a tuple of sales records.

Each record contains: (Product Name, Quantity Sold)

sales = (
    ("Laptop", 2),
    ("Mouse", 5),
    ("Laptop", 3),
    ("Keyboard", 4),
    ("Mouse", 2),
    ("Monitor", 1),
    ("Keyboard", 1),
    ("Laptop", 1)
)

Create a dictionary where:

Key = Product name
Value = Total quantity sold

Expected Output
{
    "Laptop": 6,
    "Mouse": 7,
    "Keyboard": 5,
    "Monitor": 1
}
Rules:
Do not use collections.Counter.
Do not use defaultdict.
Do not use setdefault().
Do not use dictionary comprehension.
Do not use list comprehension.
Use loops only.
Use tuple unpacking.
"""
```

```python
sales = (
    ("Laptop", 2),
    ("Mouse", 5),
    ("Laptop", 3),
    ("Keyboard", 4),
    ("Mouse", 2),
    ("Monitor", 1),
    ("Keyboard", 1),
    ("Laptop", 1)
)

product_sales = {}

for product, quantity in sales:
    if product in product_sales:
        product_sales[product] += quantity
    else:
        product_sales[product] = quantity

print(product_sales)
```

```python
"""
Problem 10 :Find the Top Scoring Student

You are given a tuple containing student records.

Each record contains: (Name, Subject, Marks)
records = (
    ("Rahul", "Math", 85),
    ("Anjali", "Math", 92),
    ("Rahul", "Science", 88),
    ("Anjali", "Science", 81),
    ("Rohit", "Math", 90),
    ("Rahul", "English", 75),
    ("Rohit", "Science", 95),
    ("Anjali", "English", 89)
)
Create a dictionary where

Key = Student name
Value = Total marks of that student

Expected dictionary:

{
    "Rahul": 248,
    "Anjali": 262,
    "Rohit": 185
}

Using that dictionary, find the student with the highest total marks.

Expected Output:

Top Student: Anjali
Total Marks: 262

Rules:
Do not use max()
Do not use sum()
Do not use dictionary comprehension
Do not use list comprehension
Do not sort
Use loops only
Use tuple unpacking
Use dictionary
"""
```

```python
records = (
    ("Rahul", "Math", 85),
    ("Anjali", "Math", 92),
    ("Rahul", "Science", 88),
    ("Anjali", "Science", 81),
    ("Rohit", "Math", 90),
    ("Rahul", "English", 75),
    ("Rohit", "Science", 95),
    ("Anjali", "English", 89)
)

student_marks = {}

for name, subject, marks in records:
    if name in student_marks:
        student_marks[name] += marks
    else:
        student_marks[name] = marks

top_student = None
highest_marks = 0

for name, total_marks in student_marks.items():
    if total_marks > highest_marks:
        highest_marks = total_marks
        top_student = name

print(f"Top Student: {top_student}")
print(f"Total Marks: {highest_marks}")
```
