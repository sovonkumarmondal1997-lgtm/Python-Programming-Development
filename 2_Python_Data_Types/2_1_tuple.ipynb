1. Creating a Tuple

```python
# 1. empty tuple:
my_tuple = ()
print(my_tuple)
print(type(my_tuple))
```

```python
# 2. creating a tuple with single item:
my_tuple = (1,)
print(my_tuple)
print(type(my_tuple))
```

```python
# homogeneous tuple:
my_tuple = (1,2,3,4,5)
print(my_tuple)
```

```python
# hetrogenious tuple:
my_tuple = (1,2.5,'hello',True)
print(my_tuple)
```

```python
# 2D tuple:
my_tuple = ((1,2,3),(2,3,4))
print(my_tuple)
```

```python
# creating tuple with type conversion:
my_tuple = tuple([1,2,3,4,5])
print(my_tuple)
print(type(my_tuple))

my_tuple_1 = tuple('hello')
print(my_tuple_1)
print(type(my_tuple_1))
```

5. Deleting Items:

```python
# Deleting a single item in a tuple is not possible but we can delete the entire tuple using del keyword:
my_tuple = (1,2,3,4,5)
print(my_tuple)
del my_tuple
print(my_tuple) # this will raise an error because the tuple is deleted
```

5. Operations on tuple:

```python
# arithmetic operations ('+'/'*') on tuples:
tuple_1 = (1,2,3)
tuple_2 = (4,5,6)
print(tuple_1 + tuple_2) # concatenation
print(tuple_1 * 3) # repetition
```

```python
# membership operator in tuple:
my_tuple = (1,2,3,4,5)
print(3 in my_tuple) # True
print(6 in my_tuple) # False

print(3 not in my_tuple) # False
print(6 not in my_tuple) # True
```

```python
# iteration in tuple:
my_tuple = (1,2,3,4,5)
for item in my_tuple:
    print(item)
```
