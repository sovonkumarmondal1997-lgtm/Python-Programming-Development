1. Dictionary Operation

```python
# membership operators in dictionary
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
print('name' in my_dict)  
print('country' in my_dict) 

print('avishek' not in my_dict)
print('sovon' not in my_dict) 
```

```python
# iterating through a dictionary using for loop
my_dict = {'name': 'avishek', 'age': 30, 'city': 'Kolkata, India'}
for key in my_dict:
    print("Key:", key)

my_dict = {'name': 'sovon', 'age': 33, 'city': 'Mumbai, India'}
for value in my_dict.values():
    print("Value:", value)

my_dict = {'name': 'sumit', 'age': 33, 'city': 'Pune, India'}
for key, value in my_dict.items():
    print("Key:", key, "Value:", value)
```

2. Dictionary Function:

```python
# length of a dictionary using len() function
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
print("Length of dictionary:", len(my_dict))
```

```python
# sorting a dictionary by keys using sorted() function
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}

sorted_dict = sorted(my_dict) # by default it sorts by keys in ascending order
print(sorted_dict)

sorted_dict = sorted(my_dict, reverse=True) # sorting by keys in descending order
print(sorted_dict)
```

```python
# items() function of dictionary
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
print(my_dict.items())
```

```python
# keys() function of dictionary
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
print(my_dict.keys())
```

```python
# values() function of dictionary
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
print(my_dict.values())
```

```python
# updating a dictionary using update() function
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
print("Original dictionary:", my_dict)
my_dict.update({'age': 25, 'country': 'USA'})
print("Dictionary after update:", my_dict)
```

3. Dictionary Comprehension:

```python
# dict = {key: value for variable in iterable if condition}
```

```python
# print first 10 numbers and their squares using dictionary comprehension
squares = {x: x**2 for x in range(1, 11)}
print(squares)
```

```python
# dictionary comprehension using a existing dictionary
my_dict = {'name': 'avishek', 'age': 33, 'city': 'New York'}
upper_dict = {key.upper(): value.upper() if isinstance(value, str) else value for key, value in my_dict.items()}
print(upper_dict)
```

```python
# dictionary comprehension using zip() function
keys = ['name', 'age', 'city']
values = ['avishek', 33, 'New York']
my_dict = {key: value for key, value in zip(keys, values)}
print(my_dict)
```

```python
# dictionary comprehension using if condition
numbers = range(1, 11)
even_squares = {x: x**2 for x in numbers if x % 2 == 0}
print(even_squares)
```

```python
# nested dictionary comprehension to create a multiplication table
multiplication_table = {i: {j: i * j for j in range(1, 11)} for i in range(1, 11)}
print(multiplication_table)
```
