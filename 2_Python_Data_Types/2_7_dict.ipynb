```python
"""
Dictionary:
-> Dictionary in Python is a collection of keys values, used to store data values like a map, which,
   unlike other data types which hold only a single value as an element.

In some languages it is known as map or assosiative arrays.

dict = { 'name' : 'sovon mondal' , 'age' : 33 , 'gender' : 'male' }

Characterstics:
1. Mutable
2. Indexing has no meaning
3. keys can't be duplicated
4. keys can't be mutable items
"""
```

1. Creating Dictionary:

```python
# create an empty dictionary
my_dict = {}
print(my_dict)
print(type(my_dict))
```

```python
# create a 1D dictionary
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print(my_dict)
```

```python
# create a homogeneous dictionary
my_dict = {'name': 'avishek', 'gender': 'male'}
print(my_dict)
```

```python
# create a dictionary with mixed keys
my_dict = {'name': 'avishek', 1: [1, 2, 3], 'age': 33}
print(my_dict)
```

```python
# create a 2D dictionary
my_dict = {'name': 'avishek', 'age': 33, 'address': {'street': '123 Main St', 'city': 'New York'}}
print(my_dict)
```

```python
# create a 3D dictionary in multiple lines

my_dict = {
    'name': 'avishek',
    'age': 33,
    'address': {
        'street': '123 Main St',
        'city': 'New York',
        'country': 'USA',
        'coordinates': {
            'lat': 40.7128,
            'lon': -74.0060
        }
    }
}
print(my_dict)
```

```python
# creating a dictionary using sequence of items and dict() constructor
my_dict = dict([('name', 'avishek'), ('age', 33), ('gender', 'male')])
print(my_dict)
```

```python
# creating a dictionary using duplicate keys
my_dict = {'name': 'avishek', 'age': 33, 'name': 'nitish'}
print(my_dict)
```

```python
# creating a dictionary using mutable items as keys are not allowed in dictionary
# my_dict = {[1, 2, 3]: 'numbers', 'name': 'avishek'}
# print(my_dict) # it will throw an error because list is a mutable item and can't be used as a key in dictionary
```

```python
# Accessing values from dictionary using keys:
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print(my_dict['name'])
print(my_dict['age'])

# Accessing values from dictionary using get() function:
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print(my_dict.get('name'))
print(my_dict.get('age'))
```

```python
# Accessing values from a 2D dictionary using keys
my_dict = {'name': 'avishek', 'age': 33, 'address': {'street': '123 Main St', 'city': 'New York'}}
print(my_dict['address']['street'])

# Accessing values from a 2D dictionary using get() function
my_dict = {'name': 'avishek', 'age': 33, 'address': {'street': '123 Main St', 'city': 'New York'}}
print(my_dict.get('address').get('street'))
```

```python
# adding new key value pair to dictionary using key and value:
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print(my_dict)
my_dict['address'] = '123 Main St'
print(my_dict)
```

```python
# adding new kay value pair to a 2D dictionary
my_dict = {'name': 'avishek', 'age': 33, 'address': {'street': '123 Main St', 'city': 'New York'}}
print(my_dict)
my_dict['address']['country'] = 'USA'
print(my_dict)
```

```python
# removing a key value pair from dictionary using pop() method
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print("Original dictionary:", my_dict)
my_dict.pop('age')
print("Dictionary after removing 'age':", my_dict)
```

```python
# removing last inserted key value pair from dictionary using popitem() method
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print("Original dictionary:", my_dict)
my_dict.popitem()
print("Dictionary after removing last inserted item:", my_dict)
```

```python
# delete a dictionary using del keyword
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print(my_dict)
del my_dict
# print(my_dict) # it will throw an error because my_dict is deleted
```

```python
# delete or remove a key value pair from dictionary using del keyword
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print(my_dict)
del my_dict['age']
print(my_dict)
```

```python
# clear all items from dictionary using clear() method
my_dict = {'name': 'avishek', 'age': 33, 'gender': 'male'}
print("original dictionary:", my_dict)
my_dict.clear()
print("Dictionary after clearing all items:", my_dict)
```

```python
# editing a existing value of a key in dictionary
my_dict = {'name': 'avishek', 'age': 33, 'address': {'street': '123 Main St', 'city': 'New York'}}
print("Original dictionary:")
print(my_dict)

my_dict['age'] = 25
print("Dictionary after update age:")
print(my_dict)
```
