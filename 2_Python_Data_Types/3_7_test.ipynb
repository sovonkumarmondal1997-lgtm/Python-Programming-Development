```python
"""
9. Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate
from the list which is closest in terms of distance from the query point.

List of Coordinates:
[(1,1),(2,2),(3,3),(4,4)]

Query Point:
(0,0)

Output:
Nearest to (0,0) is (1,1)
"""
```

```python
def shortest_distance(coordinates, query_point):
    temp_var = []

    for i in coordinates:
        distance = ((i[0] - query_point[0]) ** 2 + (i[1] - query_point[1]) ** 2) ** 0.5
        temp_var.append(distance)
    
    return coordinates[temp_var.index(min(temp_var))]

Coordinates = [(1,1),(2,2),(3,3),(4,4)]
query_point = (0,0)

shortest_distance(Coordinates, query_point)
```

```python
"""
10. Write a python program that receives a list of strings and performs bag of word operation on those strings:
https://en.wikipedia.org/wiki/Bag-of-words_model
"""
```

```python
def bow(L):

    vocab = set()

    for i in L:
        vocab.update(i.split())
    
    result = []

    for i in L:
        result.append([])
        for j in vocab:
            result[-1].append(i.count(j))

    print(vocab)
    return result

L = ['cat mat rat cat', 'sat bat cat mat rat', 'pat mat rat cat', 'cat mat rat pat']
bow(L)
```

```python
# 11. Write a Python program to add three given lists using Python map and lambda.
"""
L1 = [1,2,3]
L2 = [4,5,6]
L3 = [7,8,9]

Output:
[12, 15, 18]
"""
```

```python
L1 = [1,2,3]
L2 = [4,5,6]
L3 = [7,8,9]

list(map(lambda x,y,z: x+y+z, L1, L2, L3))
```

```python
"""
12. Write a Python program to create a list containing the power of said number in bases raised to the corresponding number
in the index using Python map.
Input:
list1 = [1,2,3,4,5,6]

Output:
[1,2,9,64,625,-]
"""
```

```python
"""
14. Use reduce to convert a 2D list to 1D list.
"""
ini_list = [[1,2,3],
            [4,5,6],
            [7,8,9]]

import functools
functools.reduce(lambda x,y: x+y, ini_list)
```
