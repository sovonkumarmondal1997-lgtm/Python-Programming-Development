```python
# np.argmax() is a function in NumPy that returns the index of the maximum element in an array along a specified axis.
```

```python
import numpy as np

a = np.random.randint(1,100,15)
print(a)

b = np.random.randint(1,100,24).reshape(6,4)
print(b)
```

```python
np.argmax(a)
```

```python
np.argmax(b)
```

```python
np.argmax(b, axis=0)
```

```python
np.argmax(b, axis=1)
```
