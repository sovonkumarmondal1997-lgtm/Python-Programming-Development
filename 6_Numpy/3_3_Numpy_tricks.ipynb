```python
# np.histogram() is a function in NumPy that computes the frequency distribution of array elements by grouping them into specified bins.
```

```python
import numpy as np

a = np.random.randint(1,100,15)
print(a)

b = np.random.randint(1,100,24).reshape(6,4)
print(b)
```

```python
np.histogram(a,bins=[0,20,40,60,80,100])
```

```python
np.histogram(b,bins=[0,20,40,60,80,100])
```

```python
# np.corrcoef() is a function in NumPy that returns the correlation coefficient matrix, showing the strength and direction of the linear relationship between variables.
```

```python
salary = np.array([20000,40000,25000,35000,60000])
experience = np.array([1,3,2,4,2])

print(np.corrcoef(salary,experience))
```

Set functions:

np.union1d

np.intersect1d

np.setdiff1d

np.setxor1d

np.in1d

```python
# np.union1d():
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([3, 4, 5, 6])

print(np.union1d(a, b))
```

```python
# np.intersect1d():
a = np.array([1, 2, 3, 4])
b = np.array([3, 4, 5, 6])

print(np.intersect1d(a, b))
```

```python
# np.setdiff1d():
a = np.array([1, 2, 3, 4])
b = np.array([3, 4, 5, 6])

print(np.setdiff1d(a, b))
```

```python
# np.setxor1d():
a = np.array([1, 2, 3, 4])
b = np.array([3, 4, 5, 6])

print(np.setxor1d(a, b))
```

```python
# np.isin():
a = np.array([1, 2, 3, 4])
b = np.array([3, 4, 5, 6])

print(np.isin(a, b))
```

```python
# np.clip() is a function in NumPy that limits the values in an array to a specified minimum and maximum range, replacing values outside the range with the nearest boundary value.
```

```python
aa = np.random.randint(1,100,15)
print(aa)

bb = np.random.randint(1,100,24).reshape(6,4)
print(bb)
```

```python
np.clip(aa,a_min=30, a_max=90)
```
