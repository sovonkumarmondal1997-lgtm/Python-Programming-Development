Brodcasting:

```python
# Broadcasting in NumPy is the process of automatically expanding arrays with different shapes so that element-wise operations can be performed.

# Broadcasting is needed in NumPy to perform operations on arrays of different shapes without manually resizing or copying the data.
```

```python
import numpy as np

a = np.arange(6).reshape(2, 3)
print(a)

b = np.arange(3).reshape(1, 3)
print(b)
```

```python
print(a + b)
```

Broadcasting Rules:

Broadcasting Rule 1: If two arrays have different numbers of dimensions, NumPy adds dimensions of size 1 to the beginning of the smaller array until both have the same number of dimensions.

Broadcasting Rule 2: 

If the sizes of two dimensions are different, NumPy stretches the dimension with size 1 to match the other.

If neither dimension is 1 and their sizes are different, broadcasting is not possible and NumPy raises an error.
