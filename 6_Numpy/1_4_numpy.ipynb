Array Operations:

```python
import numpy as np
a1 = np.arange(12).reshape(3,4)
print(a1)

a2 = np.arange(12,24).reshape(3,4)
print(a2)
```

```python
# Scalar operation: Scalar operation in a NumPy array is the process of applying a single value (scalar) to every element of the array simultaneously.

# arithmetic operations: all arithmetic operations in NumPy are element-wise, meaning that they are applied to each element of the array independently. For example, if you add a scalar value to a NumPy array, it will be added to each element of the array. and all arithmetic operators (+, -, *, /, //, %, **) can be used for scalar operations in NumPy arrays.
```

```python
a1 + 10
```

```python
a1 * 2
```

```python
a1 / 3
```

```python
# relational operations:Relational operation in NumPy compares array elements with another array or a scalar and returns a Boolean (True or False) array.
a1 == 4
```

```python
a1 > 5
```

```python
# Vector Opertions : Vector operation in NumPy performs element-wise arithmetic between two arrays of the same shape.

# Vector arithmetic operations: Vector arithmetic operations in NumPy perform element-wise addition, subtraction, multiplication, and division between arrays.
```

```python
a1 + a2
```

```python
a1 - a2
```

```python
a1 * a2
```

Array Functions:

```python
a1 = np.random.random((3,3))
print(a1)

a1 = np.round(a1*100)
print(a1)
```

```python
# max
np.max(a1)
```

```python
# 0 -> column and 1 -> row
x = np.max(a1, axis=0)
print(x)

w = np.max(a1, axis=1)
print(w)
```

```python
# min 
np.min(a1)
```

```python
# 0 -> column and 1 -> row
x = np.min(a1, axis=0)
print(x)

w = np.min(a1, axis=1)
print(w)
```

```python
# sum
np.sum(a1)
```

```python
# product
np.prod(a1)
```

```python
# mean
np.mean(a1)
```

```python
# median 
np.median(a1)
```

```python
# standard deviation
np.std(a1)
```

```python
# variance
np.var(a1)
```

```python
# trigonometric finctions : sin, cos, tan, arcsin, arccos, arctan
np.sin(a1)
```

```python
# Dot product : Dot product in NumPy multiplies corresponding elements of two vectors or matrices and sums the results to produce a single value or a new matrix.

# A dot product is valid only when the number of columns in the first matrix equals the number of rows in the second matrix.

a2 = np.arange(12).reshape(3,4)
print(a2)

a3 = np.arange(12,24).reshape(4,3)
print(a3)

np.dot(a2,a3)
```

```python
# log
np.log(a1)
```

```python
# exponential
np.exp(a1)
```
