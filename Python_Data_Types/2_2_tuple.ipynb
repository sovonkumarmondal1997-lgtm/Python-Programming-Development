7. Tuple function:

```python
# len()
my_tuple = (1,2,3,4,5)
print(len(my_tuple))
```

```python
# min() and max()
my_tuple = (1,2,3,4,5)
print(min(my_tuple)) 
print(max(my_tuple)) 
```

```python
# sum()
my_tuple = (1,2,3,4,5)
print(sum(my_tuple))
```

```python
# sorted()
my_tuple = (5,4,3,2,1)
print(sorted(my_tuple))
print(sorted(my_tuple, reverse=True))
```

```python
# count()
my_tuple = (1,2,3,4,5,1,2,3)
print(my_tuple.count(1))
print(my_tuple.count(2))
print(my_tuple.count(3))
print(my_tuple.count(4))
print(my_tuple.count(5))
print(my_tuple.count(6)) 
```

```python
# index()
my_tuple = (1,2,3,4,5,1,2,3)
print(my_tuple.index(1))
print(my_tuple.index(2))
print(my_tuple.index(3))
print(my_tuple.index(4))
print(my_tuple.index(5))
```

```python
"""
Difference between list and tuple:
1. syntax: list uses [] while tuple uses ()
2. mutability: list is mutable while tuple is immutable
3. performance: tuple is faster than list because of immutability
4. memory: tuple uses less memory than list because of immutability
5. methods: list has more methods than tuple because of mutability
6. use cases: list is used when we need to modify the data while tuple is used when we need to protect the data
7. unpacking: tuple can be unpacked while list cannot be unpacked
8. error prone: list is more error prone than tuple because of mutability
"""
```

```python
# tuple unpacking:
a,b,c,d = (1,2,3,4)
print(a)
print(b)
print(c)
print(d)
```

```python
# tuple unpacking:
my_tuple = (1,2,3,4,5)
a,b,c,d,e = my_tuple
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
# tuple unpacking
my_tuple = (1,2,3,4,5)
a,b,*rest = my_tuple
print(a)
print(b)
print(rest)
print(type(rest))
```

```python
# zipping tuples:
names = ("Alice", "Bob", "Charlie")
ages = (25, 30, 35)
zipped = zip(names, ages)
print(list(zipped))
print(tuple(zip(names, ages)))
```
