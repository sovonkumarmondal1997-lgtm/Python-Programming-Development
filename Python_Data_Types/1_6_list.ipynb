```python
"""
Traversing a list means visiting every element in the sequence one by one,
typically using a loop, to perform an action on each item.
"""
# example:
# A simple list of data points
sensor_readings = [22.5, 25.0, 19.8, 23.4]

# Traversing the list to apply a 10% calibration offset
for reading in sensor_readings:
    calibrated_value = reading * 1.1
    print(round(calibrated_value,2))
```

```python
# in python there is 2 ways to traverse a list

# 1. itemwise
# 2. indexwise
```

```python
# itemwise traverse a list:
l = [1,2,3,4,5]

for i in l:
    print(i)
```

```python
# indexwise traverse a list:
l = [1,2,3,4,5]

for i in range(0,len(l)):
    print(i)
```

```python
"""
Zip() function:
-> The zip() function acts like a physical zipper, locking elements from multiple lists together into pairs (tuples)
   so you can process them at the same time.
"""
```

```python
# write a program to add items of 2 lists indexwise: using zip() function
L1 = [1,2,3,4,5]
L2 = [-1,-2,-3,-4,-5]

s = list(zip(L1,L2))
print(s)

x = [i+j for i,j in zip(L1,L2)]
print(x)
```
