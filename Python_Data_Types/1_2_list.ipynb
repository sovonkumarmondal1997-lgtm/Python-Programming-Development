5. Access items from a list: Indexing and Slicing

```python
# Indexing: Positive for a 1D list
L = [1,2,3,4,5]

print(L[0])
print(L[1])
print(L[2])
```

```python
# Indexing: Positive for a 2D list
L = [1,2,3,[4,5]]

print(L[0])
print(L[1])
print(L[2])
print(L[3])
print(L[3][0])
print(L[3][1])
```

```python
# Indexing: Positive for a 3D list
L = [[[1,2],[3,4]],[[5,6],[7,8]]]

print(L[0]) # [[1,2],[3,4]]
print(L[1]) # [[5,6],[7,8]]

print(L[0][0]) # [1,2]
print(L[0][1]) # [3,4]

print(L[1][0]) # [5,6]
print(L[1][1]) # [7,8]

print(L[0][0][0]) # 1
print(L[0][0][1]) # 2

print(L[0][1][0]) # 3
print(L[0][1][1]) # 4

print(L[1][0][0]) # 5
print(L[1][0][1]) # 6

print(L[1][1][0]) # 7
print(L[1][1][1]) # 8
```

```python
# Indexing: Negative for a 1D list
L = [1,2,3,4,5]
print(L[-1])
print(L[-2])
print(L[-3])
```

```python
# Indexing: Negative for a 2D list
L = [1,2,3,[4,5]]
print(L[-1])
print(L[-2])
print(L[-3])
print(L[-4])
print(L[-1][-1])
print(L[-1][-2])
```

```python
# Indexing: Negative for a 3D list
L = [[[1,2],[3,4]],[[5,6],[7,8]]]

print(L[-1])
print(L[-2])

print(L[-1][-1])
print(L[-1][-2])

print(L[-2][-1])
print(L[-2][-2])

print(L[-1][-1][-1])
print(L[-1][-1][-2])

print(L[-1][-2][-1])
print(L[-1][-2][-2])

print(L[-2][-1][-1])
print(L[-2][-1][-2])

print(L[-2][-2][-1])
print(L[-2][-2][-2])
```

```python
# Slicing: Positive
L = [1,2,3,4,5,6]

print(L[0:3])
print(L[3:])

# extended slicing (or slice notation with a step)
print(L[0::2])
```

```python
# Slicing: Negative
L = [1,2,3,4,5,6]

print(L[-3:-1])
print(L[-3:])

# extended slicing (or slice notation with a step)
print(L[-5:-1:2])

# Reverse the list:
print(L[::-1])
```
