1. Sets Operations:

```python
# union(|) method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set1 | set2
print("The union of set1 and set2 is:", result)
```

```python
# intersection(&) method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set1 & set2
print("The intersection of set1 and set2 is:", result)
```

```python
# difference(-) method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set1 - set2
print("The difference between set1 and set2 is:", result)
```

```python
# symmetric_difference(^) method in sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}
result = set1 ^ set2
print("The symmetric difference between set1 and set2 is:", result)
```

```python
# membership testing in sets
my_set = {1, 2, 3, 4, 5}
print(3 in my_set)
print(6 in my_set)

print(3 not in my_set)
print(6 not in my_set)
```

```python
# iteration over a set
my_set = {1, 2, 3, 4, 5}
for item in my_set:
    print(item)
```
