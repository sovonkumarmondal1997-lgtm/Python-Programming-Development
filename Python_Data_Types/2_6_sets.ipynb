1. Set Functions:

```python
# isdisjoint() method in sets: check if two sets have no common elements
set1 = {1, 2, 3}
set2 = {4, 5, 6}
print(set1.isdisjoint(set2))
```

```python
# issubset() method in sets: check if one set is a subset of another set
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}
print(set1.issubset(set2))
```

```python
# issuperset() method in sets: check if one set is a superset of another set
set1 = {1, 2, 3, 4, 5}
set2 = {1, 2, 3}
print(set1.issuperset(set2))
```

```python
# copy() method in sets: create a shallow copy of a set
set1 = {1, 2, 3}
set2 = set1.copy()
print(set2)
```

```python
# Frozenset: an immutable version of a set
frozen_set1 = frozenset([1, 2, 3])
print(frozen_set1)
```

```python
# which functions can be used in frozenset: union(), intersection(), difference(), symmetric_difference()
frozen_set1 = frozenset([1, 2, 3])
frozen_set2 = frozenset([3, 4, 5])
print(frozen_set1.union(frozen_set2))
print(frozen_set1.intersection(frozen_set2))
print(frozen_set1.difference(frozen_set2))
print(frozen_set1.symmetric_difference(frozen_set2))
```

```python
# usecase of frozenset: as a key in a dictionary
frozen_set1 = frozenset([1, 2, 3])
my_dict = {frozen_set1: "This is a frozenset"}
print(my_dict[frozen_set1])
```

```python
# 2D forzenset: a frozenset of frozensets
frozen_set1 = frozenset([frozenset([1, 2]), frozenset([3, 4])])
print(frozen_set1)

# 3D forzenset: a frozenset of frozensets of frozensets
frozen_set1 = frozenset([frozenset([frozenset([1, 2]), frozenset([3, 4])]), frozenset([frozenset([5, 6]), frozenset([7, 8])])])
print(frozen_set1)
```

```python
# Set comprehension: create a set using a for loop and an optional condition
squared_set = {x**2 for x in range(10)}
print(squared_set)

normal_set = {x for x in range(10)}
print(normal_set)
```

```python
# Set comprehension with conditions
even_squared_set = {x**2 for x in range(10) if x % 2 == 0}
print(even_squared_set)
```
