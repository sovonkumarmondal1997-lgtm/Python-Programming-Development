```python
"""
List Comprehension:
-> List Comprehension provides a concise way of creating lists.
-> newlist = [expression for item in iterable if condition == True]

Advantages of List Comprehension:
1. More time-efficient and space-efficient than loops.
2. Require fewer lines of code.
3. Transforms iterative statement into a formula.
"""
```

```python
# 1.0 Add 1 to 10 numbers into a list: using a loop
L = []
for i in range(1,11):
    L.append(i)

print(L)
```

```python
# 1.1 Add 1 to 10 numbers into a list: using list_comprehension
L = [i for i in range(1,11)]
print(L)
```

```python
# 2.0 Scaler multiplication on a vector: using loop
vector = [2,3,4]
scaler = -3
# output of the multiplication should be: [-6,-9,-12]

x = []
for i in vector:
    x.append(i * scaler)

print(x)
```

```python
# 2.1 Scaler multiplication on a vector: using list_comprehension
vector = [2,3,4]
scaler = -3
# output of the multiplication should be: [-6,-9,-12]

x = [scaler * i for i in vector]
print(x)
```

```python
# 3. Add squres of the list:
L = [3,5,8,7,1,5]

X = [i ** 2 for i in L]
print(X)

Y = [i * i for i in L]
print(Y)
```

```python
# 4. print all the numbers divisible by 5 in the range of 1 to 50

x = [i for i in range(1,51) if i%5 == 0]
print(x)
```

```python
# 5. find the languages which starts with letter 'p'
languages = ['java','python','php','c','javascript','pega','scala','c++','fortarn','r','perl','pony']

x = [i for i in languages if i.startswith('p')]
print(x)

y = [language for language in languages if language.startswith('p')] 
# translation: language for every language in languages if language.startswith('p')
print(y)
```

```python
# 6. Nested if with List Comprehension:
# add new list from my_fruits and if the items of my_fruits exists in basket and also starts with 'a'
basket = ['apple','guava','cherry','banana','grapes','mango','berries','avocado']
my_fruits = ['apple','kiwi','grapes','banana','avocado','acerola','cherry']

x = [i for i in my_fruits if i in basket if i.startswith('a')]
print(x)

y = [fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
print(y)
```

```python
# 7. Print a (3,3) matrix using list comprehension -> Nested List comprehension
x = [[i*j for i in range(1,4)] for j in range(1,4)]
print(x)
```
