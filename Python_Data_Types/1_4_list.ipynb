9. Functions on Lists:

```python
# len(): it will work both homogenious and hetrogenious data
L = [3,4,5,6,7,11]
print("length of list is",len(L))

L = ['hello',11,2,3,4,5,True]
print("length of list is",len(L))
```

```python
# min(): it will work only homogenious data 
L = [3,4,5,6,7,11]
print(min(L))
```

```python
# max(): it will work only homogenious data 
L = [3,4,5,6,7,11]
print(max(L))
```

```python
# sorted(): it will work only homogenious data and it is not a parmanent operation
L = [3,4,5,6,7,11]
print(sorted(L))
```

```python
# sorted(): it will work only homogenious data and it is not a parmanent operation
L = [3,4,5,6,7,11]
print(sorted(L,reverse = True))
```

```python
# sum(): it will work only homogenious data and it is not a parmanent operation
L = [3,4,5,6,7,11]
print("sum of list is", sum(L))
```

```python
# count: count the frequency of a item inside a list
M = [1,2,1,2,1,2,3,4,11,11,11,11,11,11]

print(M.count(1))
print(M.count(2))
print(M.count(11))
```

```python
# index: find the index position of the item inside a list
M = [1,2,1,2,1,2,3,4,11,11,11,11,11,8]

print(M.index(8))
print(M.index(11))
print(M.index(11))
print(M.index(11))
print(M.index(11))
print(M.index(1))
print(M.index(2))

# if the item present in the list multiple times with duplicat value index can find the first item position
# it can not identify the other positions of the same value
```

```python
# reverse(): reverse the list permanently
L = [1,8,3,4,1,2,3,4,9]
print("Original list",L)

L.reverse()
print("Reversed list", L)
```

```python
# sort(): parmanently sort the list ascending to descending
L = [2,7,1,6,9,4]
print("original list", L)
L.sort()
print("sorted list", L)
```

```python
# copy(): copy the list in the memory
L = [6,3,8,1,9,2,4,5]
print(L)

L1 = L.copy()
print(L1)
```
