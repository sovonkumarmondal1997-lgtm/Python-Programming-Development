```python
"""
Sets:
-> A set is an unordered collection of items. Every set element is unique (no duplicates) and
   must be immutable (cannot be changed). However, a set itself is mutable. We can add or remove items from it.

-> Basically, sets are mutable but their elements are immutable.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
1. Unordered
2. Mutable
3. No Duplicates
4. Can't contain mutable data types

"""
```

1. Creating Sets:

```python
"""
1. Set's items are immutable so we can not create 2D sets or sets of lists or sets of dictionaries,
   We can only create sets of immutable data types like int, float, string, tuple, etc.
2. We can create sets using curly braces {} or using the set() constructor.
   But {} is not alloweed for creating an empty set because it creates an empty dictionary.
3. Set's items can not contain duplicate values. If we try to create a set with duplicate values,
   it will automatically remove the duplicates and keep only one instance of each value.

"""
```

```python
# creating an empty set
my_set = set()
print(my_set)
print(type(my_set))
```

```python
# creating a 1D and homogeneous set
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
# creating a hetrogenious set
my_set = {1, 2.5, 'Hello', (1, 2),True}
print(my_set)
```

```python
# creating a set using type conversion
my_list = [1, 2, 3, 4, 5]
print(my_list)
print(type(my_list))

my_set = set(my_list)
print(my_set)
print(type(my_set))
```

```python
"""
1. Sets are unordered, so the items in a set do not have a defined order.
   This means that we cannot access items in a set using an indexing() or slicing() like we do with lists or tuples.
   We can only check for the presence of an item in a set or iterate through the items in a set.
# Accessing items and Editing items not alloweed in sets.
2. Sets are mutable, which means we can add or remove items from a set after it has been created.
# Adding items and deleting or removing items from a set is alloweed.
"""
```

```python
# Adding single item to a set using add() function
my_set = {1, 2, 3}
print("Original set: ", my_set)

my_set.add(4)
print("Set after adding 4: ", my_set)
```

```python
# Adding multiple items to a set using update() function
my_set = {1, 2, 3}
print("Original set: ", my_set)
my_set.update([4, 5, 6])
print("Set after adding multiple items: ", my_set)
```

```python
# Deleting or removing particular items from a set using remove() function
my_set = {1, 2, 3, 4, 5}
print(my_set)

my_set.remove(3)
print(my_set)

# using 'del' keyword we can delete the entire set
```

```python
# Deleting or removing a particular item from a set using discard() method
my_set = {1, 2, 3, 4, 5}
print("Original set: ", my_set)

my_set.discard(3)
print("Set after discarding 3: ", my_set)
```

```python
# Delete or Remove any random item from a set using pop() method
my_set = {1, 2, 3, 4, 5}
print("Original set: ", my_set)

my_set.pop()
print("Set after popping an item: ", my_set)
```

```python
# Clearing all items from a set using clear() method
my_set = {1, 2, 3, 4, 5}
print("Original set: ", my_set)

my_set.clear()
print("Set after clearing all items: ", my_set)
```
