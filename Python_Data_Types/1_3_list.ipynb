6. Editing (Adding, Update or Edit) items in a list:

```python
# Adding items in a list: append (add single item into the list at the end)
L = [1,2,3,4,5]
L.append(6)
print(L)

# Adding a single list using append
M = [1,2,3,4,5]
M.append([4,5,6,7])
print(M)
```

```python
# Adding items in a list: extend (add multiple items into the list at the end)
L = [1,2,3,4,5,6]
L.extend([7,8,9])
print(L)

# Adding multiple items using extend 
M = [1,2,3,4,5]
M.extend('kolkata knight riders')
print(M)

# Adding single items using extend 
N = [1,2,3,4,5]
N.extend(['kolkata knight riders'])
print(N)
```

```python
# Adding items in a list: insert (add item into list at any desired location)
L = [1,2,3,4,5]
L.insert(1,100)
print(L)
```

```python
# Editing or Updating single item in a list: Negative indexing
L = [1,2,3,4,5]
print("Original list:", L)
L[-1] = 100
L[-3] = 500
print("Updated list:", L)
```

```python
# Editing or Updating single item in a list: Positive indexing
L = [1,2,3,4,5]
print("Original list:", L)
L[0] = 100
L[3] = 500
print("Updated list:", L)
```

```python
# Editing or Updating multiple items in a list: Positive Slicing
L = [1,2,3,4,5]
print("Original list:", L)
L[1:4] = [100,200,300]
print("Updated list:", L)
```

```python
# Editing or Updating multiple items in a list: Negative Slicing
L = [1,2,3,4,5]
print("Original list:", L)
L[-5:-1] = [100,200,300]
print("Updated list:", L)

M = [1,2,3,4,5]
print("Original list:", M)
M[:-1] = [100,200,300]
print("Updated list:", M)
```

```python
# Editing or Updating multiple items in a list: Hybrid Slicing
L = [1,2,3,4,5]
print("Original list:", L)
L[1:-1] = [100,200,300]
print("Updated list:", L)
```

7. Deleting items from a List

```python
# del: delete the entire list
L = [5,4,3,2,1]
print(L)

del L
print(L)
```

```python
# del: delete single item from the list
M = [5,4,3,2,1]
print("Original list before deletion:", M)

# del: delete single item from the list by negative indexing
del M[-1]
print("Updated list after deleting single items:", M)

# del: delete single item from the list by positive indexing
del M[0]
print("Updated list after deleting single items:", M)
```

```python
# del: delete multiple items from the list by negative indexing 
L = [1,2,3,4,5]
print("Original list:", L)
del L[-4:-1]
print("Updated list:", L)

# del: delete multiple items from the list by positive indexing
L = [1,2,3,4,5,6]
print("Original list:", L)
del L[1:5]
print("Updated list:", L)
```

```python
# remove: delete or remove the item from the list based on the value
A = [1,2,3,4,5,60]
print("Original list:", A)

A.remove(60)
print("Updated list:", A)
```

```python
# pop
A = [1,2,3,4,5,60]
print("Original list:", A)

# deleting value by the index position using pop
A.pop(0)
print("Updated list:", A)

# delete last item from the list using pop
B = [1,2,3,4,5,60]
print("Original list:", B)

B.pop()
print("Updated list:", B)

```

```python
# clear: clear the list items
L = [5,8,9,3,5,4,3,1,10]
print("Original list:", L)

L.clear()
print("Updated list:", L)
```

8. Operations on Lists:

a. Arithmetic Operations
b. Membership Operations
c. Loop Operations

```python
# Arithmetic Operations: '+'
L1 = [1,2,3,4]
L2 = [5,6,7,8]
print(L1 + L2)
```

```python
# Arithmetic Operations: '*'
L1 = [1,2,3,4]
print(L1 * 3)
```

```python
# Membership Operations: 'in'/'not in'
L1 = [1,2,3,4,5]
L2 = [1,6,7,8,[5,6]]

print(5 in L1)
print(5 in L2)

print(5 not in L1)
print(5 not in L2)

```

```python
# Loop Operations:
L1 = [1,2,3,4,5]
L2 = [1,6,7,8,[5,6]]

for i in L1:
    print(i)
```

```python
# Loop Operations:
L1 = [1,2,3,4,5]
L2 = [1,6,7,8,[5,6]]

for i in L2:
    print(i)
```

```python
# Loop Operations:
L = [[[1,2],[3,4]],[[5,6],[7,8]]]

for i in L:
    print(i)
```
