```python
# What is a Higher-Order Function in Python?
#
# A higher-order function is a function that does one of two things:
# 1. Takes another function as an argument (input)
# 2. Returns a function as its result (output)
#
# In simple terms, it's a function that works with other functions.
# This is possible because in Python, functions are "first-class citizens" -
# they can be passed around just like any other value (like numbers or strings).
#
# Why are they useful?
# - They allow for more flexible and reusable code.
# - They enable functional programming patterns.
# - They can make code more concise and expressive.
#
# Examples:
# - map(), filter(), and reduce() are built-in higher-order functions.
# - Decorators are also a form of higher-order functions.
```

```python
# Simple example:
def apply_operation(func, x, y):
    """
    This function takes another function (func) as an argument
    and applies it to two values (x and y).
    """
    return func(x, y)

def add(a, b):
    """
    This function adds two numbers and returns the result.
    """
    return a + b

def multiply(a, b):
    """
    This function multiplies two numbers and returns the result.
    """
    return a * b

# Using the higher-order function
result1 = apply_operation(add, 3, 4)  # Returns 7
result2 = apply_operation(multiply, 3, 4)  # Returns 12

print("Addition result:", result1)
print("Multiplication result:", result2)
```

```python
# Example 1: Function that takes another function as an argument
def apply_operation(func, x, y):
    """
    This function takes another function (func) as an argument
    and applies it to two values (x and y).
    """
    return func(x, y)

def add(a, b):
    """
    This function adds two numbers and returns the result.
    """
    return a + b

def multiply(a, b):
    """
    This function multiplies two numbers and returns the result.
    """
    return a * b

# Using the higher-order function
result1 = apply_operation(add, 3, 4)  # Returns 7
result2 = apply_operation(multiply, 3, 4)  # Returns 12

print("Addition result:", result1)
print("Multiplication result:", result2)
```

```python
# Example 2: Function that returns another function
def create_multiplier(factor):
    """
    This function returns a function that multiplies a number by a given factor.
    """
    def multiplier(x):
        """
        This inner function multiplies a number (x) by the factor.
        """
        return x * factor
    return multiplier

# Using the function that returns a function
double = create_multiplier(2)  # double is now a function that multiplies by 2
triple = create_multiplier(3)  # triple is now a function that multiplies by 3

print("Double 5:", double(5))  # Returns 10
print("Triple 5:", triple(5))  # Returns 15
```

```python
# Explaining the map() function in simple terms:
#
# The map() function is a built-in higher-order function in Python.
# "Higher-order" means it can take other functions as arguments or return functions.
#
# What does map() do?
# - It applies a given function to each item in a collection (like a list, tuple, etc.)
# - It returns a map object (which is an iterator) containing the results
# - You usually convert it to a list to see the results
#
# Syntax: map(function, iterable)
# - function: The function to apply to each item
# - iterable: The collection of items to process
#
# Why is it useful?
# - It allows you to transform data in a list without using loops
# - It's more concise and often faster than traditional loops
# - It promotes functional programming style
#
# Simple example:
# numbers = [1, 2, 3, 4, 5]
# squared = list(map(lambda x: x**2, numbers))
# This squares each number: [1, 4, 9, 16, 25]
#
# You can also use named functions instead of lambda:
# def square(x):
#     return x**2
# squared = list(map(square, numbers))
#
# Note: In Python 3, map() returns an iterator, not a list.
# So we wrap it with list() to get the actual list of results.
```

```python
# Simple higher-order function using map to square items in a list

def square_list_items(my_list):
    """
    This is a higher-order function that uses map
    The lambda function squares each item
    """
    return list(map(lambda x: x ** 2, my_list))

# Example usage
my_list = [1, 2, 3, 4, 5]
squared_list = square_list_items(my_list)

print("Original list:", my_list)
print("Squared list:", squared_list)
```

```python
# Simple higher-order function using map to label items in a list as odd or even

def label_odd_even(my_list):
    # This is a higher-order function that uses map
    # The lambda function checks if each number is even or odd
    return list(map(lambda x: "even" if x % 2 == 0 else "odd", my_list))

# Example usage
my_list = [1, 2, 3, 4, 5, 6]
labeled_list = label_odd_even(my_list)

print("Original list:", my_list)
print("Labeled list:", labeled_list)
```

```python
# Simple higher-order function using map to extract names from a list of user dictionaries

def extract_names(users):
    # This is a higher-order function that uses map
    # The lambda function extracts the 'name' key from each dictionary
    return list(map(lambda user: user["name"], users))

# Example usage with 5 users
users = [
    {"name": "Alice", "age": 25, "gender": "Female"},
    {"name": "Bob", "age": 30, "gender": "Male"},
    {"name": "Charlie", "age": 35, "gender": "Male"},
    {"name": "Diana", "age": 28, "gender": "Female"},
    {"name": "Eve", "age": 22, "gender": "Female"}
]

names_list = extract_names(users)

print("User data:", users)
print("Extracted names:", names_list)
```
