```python
# How functions executed in memory:
# 1. When a function is called, Python creates a new frame in memory to execute the function's code.
# 2. This frame contains the function's local variables and parameters.
# 3. Once the function finishes executing, the frame is removed from memory, 
# and the function's local variables are no longer accessible.
```

```python
# without return statement, the function will return None by default

# example of a function without a return statement:
def greet(name):
    """
    This function takes a name as an argument and prints a greeting message.
    """
    print(f"Hello, {name}!")

greeting = greet("Alice")
print(greeting)  
```

```python
# variable scope in python:
# 1. Local scope: Variables defined inside a function are in the local scope and can only be accessed within that function.
# 2. Global scope: Variables defined outside of any function are in the global scope and can be accessed from anywhere in the code.
# 3. Enclosing scope: Variables defined in an enclosing function can be accessed by nested functions, but not by the global scope.
```

```python
# example of local scope and access within that function:
def outer_function():
    outer_variable = "I am an outer variable" # This variable is defined in the outer function and can be accessed by the inner function
    
    def inner_function():
        inner_variable = "I am an inner variable" # This variable is defined in the inner function and can only be accessed within the inner function
        print(outer_variable)  # Accessing outer variable from inner function
        print(inner_variable)  # Accessing inner variable from inner function
    
    inner_function() # Calling the inner function to execute its code
    # print(inner_variable)  # This will raise an error because inner_variable is not accessible in the outer function
    print(outer_variable)  # This will work because outer_variable is in the global scope
outer_function() # Calling the outer function to execute its code
```

```python
# example of global scope and access from anywhere in the code:
global_variable = "I am a global variable" # This variable is defined in the global scope and can be accessed from anywhere in the code

def access_global_variable():
    """
    This function accesses the global variable and prints its value.
    """
    print(global_variable)  # Accessing global variable from within a function  
access_global_variable()  # Calling the function to access the global variable
print("Accessing global variable from outside the function:", global_variable)  # Accessing global variable from outside the function
```

```python
# example of enclosing scope and access from nested functions:
def outer_function():
    """
    This function demonstrates the concept of enclosing scope by defining an inner function that accesses a variable from the outer function.
    """
    outer_variable = "I am an outer variable" # This variable is defined in the outer function and can be accessed by the inner function
    
    def inner_function():
        """
        This inner function accesses the outer variable and prints its value.
        """
        print(outer_variable)  # Accessing outer variable from inner function
    
    inner_function() # Calling the inner function to execute its code
outer_function() # Calling the outer function to execute its code
```

```python
# 1. main program scope: 
# This is the top-level scope where the main program code is executed. 
# Variables defined in this scope can be accessed from anywhere in the code, including within functions.

# simple example of main program scope in a function
main_variable = "I am a main program variable"

def my_function():
    """
    This function demonstrates the concept of main program scope by accessing a variable defined in the main program scope.
    """
    print(main_variable)  # Accessing main program variable from within a function

my_function()

# 2. function scope:
# This is the scope created when a function is called.
# Variables defined in this scope are local to the function and cannot be accessed outside of it.

# simple example of function scope
def my_function():
    """
    This function demonstrates the concept of function scope by defining a variable in the function scope and accessing it from within the function.
    """
    function_variable = "I am a function variable"  # This variable is defined in the function scope
    print(function_variable)  # Accessing function variable from within the function

my_function()
```

```python
# creating a function using same variable name in local variable and global variable and see the difference in output
global_variable = "I am a global variable"
def my_function():
    """
    This function demonstrates the concept of local scope by defining a variable in the local scope and accessing it from within the function.
    """
    global_variable = "I am a local variable"  # This variable is defined in the function scope and has the same name as the global variable
    print(global_variable)  # Accessing local variable from within the function
my_function()  # This will print the local variable, not the global variable
print(global_variable)  # This will print the global variable, not the local variable
```

```python
# example of nested functions:
def outer_function():
    outer_variable = "I am an outer variable"
    
    def inner_function():
        inner_variable = "I am an inner variable"
        print(outer_variable)  # Accessing outer variable from inner function
        print(inner_variable)  # Accessing inner variable from inner function
    
    inner_function() # Calling the inner function to execute its code
    # print(inner_variable) # This will raise an error because inner_variable is not accessible in the outer function
    print("Accessing outer variable from outer function:", outer_variable) # This will work because outer_variable is in the global scope

outer_function()
```

```python
# Functions are 1st class objects in Python, which means they can be treated like any other object.
# This means that functions can be assigned to variables, passed as arguments to other functions,
# and returned from other functions.

# example of functions being treated as first-class objects:
def greet(name):
    """
    This function takes a name as an argument and returns a greeting message.
    """
    return f"Hello, {name}!"

# Assigning a function to a variable
greeting_function = greet

# Calling the function through the variable
print(greeting_function("Alice"))
```

```python
# using type and id to check the type and memory address of a function
print(type(greet))  # This will show that greet is a function
print(id(greet))  # This will show the memory address of the greet function
```

```python
# reassigning the function to a new variable and checking the type and id again
new_greet = greet  # Reassigning the greet function to a new variable
print(type(new_greet))  # This will show that new_greet is also a function
print(id(new_greet))  # This will show the same memory address as greet, since new_greet is just another reference to the same function 
```

```python
# delete a function name using del keyword and see the effect on the function object

def say_hello():
    print("Hello!")

say_hello()  # works
```

```python
del say_hello
```

```python
say_hello()  # This will raise a NameError because say_hello has been deleted and is no longer defined in the current scope
```

```python
# Demonstrating that functions are first-class objects in Python
# Storing the say_hello function in a list with integers

my_list = [1, 2, say_hello, 4, 5]

# Check if it works by calling the function from the list
print("Calling say_hello from the list:")
my_list[2]()  # This should print "Hello!"

print("The list contains:", my_list)    
```

```python
# Checking if say_hello function is mutable or immutable using sets
# Functions are immutable objects in Python, so they can be stored in sets

# Since say_hello was deleted, but it's still in my_list, we can get it from there
func_from_list = my_list[2]

# Try to add the function to a set
my_set = {1, 2, func_from_list, 4, 5}

print("Set containing the function:", my_set)
print("Function in set:", func_from_list in my_set)

# Call the function from the set
print("Calling function from set:")
list(my_set)[2]()  # Since sets are unordered, but we know the function is there
```

```python
# Creating two functions where one returns another function

def create_greeter(greeting):
    """
    This function creates a function that prints a greeting message.
    """
    def greeter(name):
        """
        This function takes a name as a parameter and prints a greeting message.
        """
        print(f"{greeting}, {name}!")
    return greeter

# Create a greeter function
hello_greeter = create_greeter("Hello")

# Call the returned function
hello_greeter("Alice")
hello_greeter("Bob")

```

```python
# Another example where one returns another function
def outer_function():
    """
    This function creates a function that prints a message.
    """
    def inner_function():
        """
        This function prints a message.
        """
        print("This is the inner function being called!")
    return inner_function

# Get the inner function
returned_func = outer_function()

# Call the returned function
returned_func()
```

```python
# Creating two functions where one function is passed as an argument to another function

def apply_function(func, value):
    """
    This function applies a function to a value and returns the result.
    """
    return func(value)

def square(x):
    """
    This function returns the square of a number.
    """
    return x * x

def cube(x):
    """
    This function returns the cube of a number.
    """
    return x * x * x

# Pass square function as argument
result1 = apply_function(square, 5)
print(f"Square of 5: {result1}")

# Pass cube function as argument
result2 = apply_function(cube, 3)
print(f"Cube of 3: {result2}")
```

```python
# Another example where one function is passed as an argument to another function
def greet_person(func, name):
    """
    This function takes a function and a name as arguments and returns the result of calling the function with the name.
    """
    return func(name)

def formal_greeting(name):
    """
    This function returns a formal greeting message.
    """
    return f"Good day, {name}!"

def casual_greeting(name):
    """
    This function returns a casual greeting message.
    """
    return f"Hey, {name}!"

print(greet_person(formal_greeting, "Alice"))
print(greet_person(casual_greeting, "Bob"))
```
