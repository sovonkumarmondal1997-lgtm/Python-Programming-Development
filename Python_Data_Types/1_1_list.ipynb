```python
"""
2. Array vs List:

1. Arrays are fixed (before store any data in array we have to define how many data we are going to store) and
   Lists are dynamic (we can add on or store more data on the go)
2. Arrays are homogenious (we can store one type of data in a array at a time) and 
   Lists are hetrogenious (we can store different types of data in a lists at a time)
3. Arrays are fast in execution and lists are slow in execution
4. Arrys are occupied less memory than lists and lists are occupied more memory than arrays for a particular work
"""
```

```python
"""
How array and list store in the memory address:

Java (array = [1,2,3]):
In Java, the array is stored like a straight row of boxes in memory. 
The variable name holds the starting address of that row. Inside the row,
the numbers 1, 2, and 3 are stored directly next to each other in continuous memory.
Because they are in order and fixed size, Java can quickly find any value using its position.

Python (list = [1,2,3]):
In Python, the list is like a row of boxes that store addresses, not the actual numbers.
Each box points to a separate place in memory where the numbers 1, 2, and 3 are stored as objects.
So the list stores references, not values directly, which makes it more flexible but slightly less efficient.

"""
# create a list and store it in a vaiable
L = [1,2,3]

# print the list address
print(id(L))

# Print the memeory address of the items [1,2,3] in the list
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))

# Print the memory address of the itrms [1,2,3]
print(id(1))
print(id(2))
print(id(3))

# here we can see memory adress of 1,2,3 and the memory adress of [1,2,3] inside the list is same
# it means in python list dose not store the items it stores the memeory address of the items
```
