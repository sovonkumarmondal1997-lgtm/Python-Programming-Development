Pandas Series:

- A pandas Series is a one-dimensional labeled array that stores a sequence of values with an associated index.

```python
import numpy as np
import pandas as pd
```

Series from lists:

```python
# string:
country = ['India', 'Pakistan', 'Nepal', 'Sri Lanka', 'Bhutan', 'Maldives', 'Afghanistan', 'Bangladesh']
pd.Series(country)
```

```python
# integer:
age = [10, 20, 30, 40, 50, 60, 70, 80]
age11 =pd.Series(age)
print(age11)
```

```python
# custom index:
country = ['India', 'Pakistan', 'Nepal', 'Sri Lanka', 'Bhutan', 'Maldives', 'Afghanistan', 'Bangladesh']
age = [10, 20, 30, 40, 50, 60, 70, 80]
pd.Series(data=age, index=country)
```

```python
# setting a name:
pd.Series(data=age, index=country, name='Country Age')
```

Series from dictionary:

```python
marks = {
    'math': 88,
    'science': 90,
    'english': 92,
    'social': 91,
    'hindi': 93,
    'sanskrit': 94
}
```

```python
marks_series = pd.Series(marks, name='Subject_Marks')
print(marks_series)
```
