Series Indexing:

```python
# integer indexing:
x = pd.Series([12,13,14,35,56.46,79,9])
print(x[0])
print(x[1])
print(x[2])
```

```python
# Series only work with positive indexing not work with negative indexing
```

Editing Series:

```python
marks = {
    'math': 88,
    'science': 90,
    'english': 92,
    'social': 91,
    'hindi': 93,
    'sanskrit': 94
}

marks_series = pd.Series(marks, name='Subject_Marks')
print(marks_series)
```

```python
# using indexing:
print(marks_series)
```

```python
marks_series[1] = 100
print(marks_series)
```

```python
marks_series['botany'] = 100
print(marks_series)
```

```python
# slicing:
vk[2:4] = [88,99]
print(vk)
```

```python
# fancy indexing:
vk[[1,2,3,4,5]] = [88,99,100,101,102]
print(vk)
```

```python
# using index labels:
movies['Uri: The Surgical Strike'] = 'Roman Reigns'
print(movies)
```
