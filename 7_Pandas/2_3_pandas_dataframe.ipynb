```python
# how many super overs were played in ipl
mask = ipl['SuperOver'] == 'Y'
ipl[mask].shape[0]
```

```python
# how many matches csk has won in kolkata
mask = (ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')
ipl[mask].shape[0]
```

```python
# toss winner is match winner in percentage
mask = ipl['WinningTeam'] == ipl['TossWinner']
(ipl[mask].shape[0]/ipl.shape[0])*100
```

```python
# movies with rating higher than 8 and votes > 10000
mask = (movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)
movies[mask].shape[0]
```
