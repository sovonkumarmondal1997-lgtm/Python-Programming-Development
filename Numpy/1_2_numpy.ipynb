```python
# creating numpy arrays:
```

```python
# 1D np.array or vector
import numpy as np
a = np.array([1, 2, 3])
print(a)
print(type(a))
```

```python
# 2D np.array or matrix
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)
print(type(b))
```

```python
# 3D np.array or tensor
c = np.array([[[1, 2], [4, 5]], [[7, 8], [10, 11]]])
print(c)
print(type(c))
```

```python
# data types in numpy arrays
d = np.array([1, 2, 3], dtype=float)
print(d)
print(type(d))
```

```python
# np.arange()
a = np.arange(0, 10)
print(a)

b = np.arange(0, 10, 2)
print(b)
```

```python
# with reshape() functon:
np.arange(0, 100, 10).reshape(2, 5) #first is rows and second is columns
```

```python
# np.ones and np.zeros
np.ones((2, 3)) # 2 rows and 3 columns of ones
```

```python
np.zeros((2, 3)) # 2 rows and 3 columns of zeros
```

```python
# np.random
b = np.random.rand(3, 4) # 3 rows and 4 columns of random numbers between 0 and 1
print(b)

c = np.random.random((3, 4)) # 3 rows and 4 columns of random numbers between 0 and 1
print(c)
```

```python
# np.linspace()
a = np.linspace(0, 10, 5) # start, stop, number of points
print(a)

b = np.linspace(-10, 10, 10) # start, stop, number of points
print(b)
```

```python
# np.identity()
np.identity(3) # 5x5 identity matrix
```
