Numpy array vs Python list:

```python
# speed view:
# NumPy is faster than Python lists because it stores data in contiguous memory and performs optimized vectorized operations using compiled C code.
```

```python
# memeory or space view:
# Python lists consume more memory because they store references to objects rather than the actual values. In contrast, NumPy arrays store elements of the same data type in contiguous memory, making them much more memory-efficient. Additionally, NumPy allows you to control the size of each element using data types such as int8, int16, int32, and int64. By choosing a smaller data type when appropriate, you can further reduce memory usage and improve performance, especially when working with large datasets.
```

Advanced Indexing:

```python
# fancy indexing:
import numpy as np
a = np.arange(24).reshape(6,4)
print(a)
```

```python
a[[0,2,3]]
```

```python
a[[0,2,3,5]]
```

```python
a[:,[0,2,3]]
```

```python
# Boolean Indexing:
a = np.random.randint(1,100,24).reshape(6,4)
print(a)

print(a>50)
```

```python
# find all numbers grater than 50:
a[a>50]
```

```python
# find out even number in array:
a[a%2==0]
```

```python
# find all the numbers grater than 50 and even:
a[(a>50) & (a%2==0)]
```

```python
# find all the numbers divisible by 7:
a[a%7==0]
```
