indexing and slicing:

```python
# Indexing:
```

```python
import numpy as np
a1 = np.arange(10)
print(a1)

a2 = np.arange(12).reshape(3,4)
print(a2)

a3 = np.arange(8).reshape(2,2,2)
print(a3)
```

```python
# find the last item from the array: through indexing 
print(a1[-1])
print(a1[0])
```

```python
print(a2)

# fetch 6 from the array:
print(a2[1][2])
print(a2[1,2])
```

```python
# fetch item from a 3D array:
print(a3)
print(a3[1][1][1])
print(a3[1,1,1])
```

```python
# Slicing:
```

```python
print(a1)
```

```python
# find 2,3,4 items from the array:
print(a1[2:5])
```

```python
print(a2)
```

```python
# fetech 1st row from a2:
print(a2[0,:])
```

```python
# fetch 2nd column from a2:
print(a2[:,2])
```

```python
# fetch 5,6 or 9,10 from a2:
print(a2[1:,1:3])
```

```python
# fetch 0,3,8,11 from a2:
print(a2[::2,::3])
```

```python
a4 = np.arange(27).reshape(3,3,3)
print(a4)
```

```python
"""
 fetch below matrix from a4:
 [[ 9 10 11]
  [12 13 14]
  [15 16 17]]
"""
print(a4[1])
```

```python
print(a4[::2])
```

```python
print(a4[0,1,:])
```

```python
print(a4[1,:,1])
```

```python
print(a4[2,1:,1:])
```

```python
print(a4[::2,0,::2])
```

iterating:

```python
print(a1)
for i in a1:
    print(i)
```

```python
print(a2)

for i in a2:
    print(i)
```

```python
print(a4)

for i in a4:
    print(i)
```

```python
for i in np.nditer(a4):
    print(i)
```

Reshaping:

```python
# transpose of a2:
print(a2)
print(np.transpose(a2))
```

```python
print(a2)
print(a2.T)
```

```python
# ravel
print(a2)
print(np.ravel(a2))
```

```python
print(a4)
print(a4.ravel())
```

Stacking:

```python
# horizontal stack
aa = np.arange(12).reshape(3,4)
print (aa)

ab = np.arange(12,24).reshape(3,4)
print (ab)
```

```python
np.hstack((aa,ab))
```

```python
# vertical stack
np.vstack((aa,ab))
```

Splitting:

```python
# horizontal spliting:
np.vsplit(aa,3)
```

```python
# vertical spliting:
np.vsplit(aa,3)
```
