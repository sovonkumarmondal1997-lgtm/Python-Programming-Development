Array Attributes:

```python
import numpy as np
a1 = np.arange(10,dtype=np.int32)
print(a1)
```

```python
a2 = np.arange(12,dtype=float).reshape(3,4)
print(a2)
```

```python
a3 = np.arange(8).reshape(2,2,2)
print(a3)
```

```python
# ndim or no. of dimensions
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
# shape or no. of elements in each dimension
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
# size or total no. of elements in the array
print(a1.size)
print(a2.size)
print(a3.size)
```

```python
# itemsize or size in bytes of each element in the array
print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```

```python
# dtype or data type of the elements in the array
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

changing datatype:

```python
a3.dtype
```

```python
# astype() method to change the data type of the array
a3 = a3.astype(np.int32)
print(a3.dtype)
```
