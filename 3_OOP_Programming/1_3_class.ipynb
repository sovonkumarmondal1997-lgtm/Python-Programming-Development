```python
"""
1. What is class diagram?
-> A class diagram is a visual representation in OOP that shows classes, their attributes, methods, and relationships between classes.

->In a class diagram, + means public access (accessible from anywhere) and - means private access (accessible only inside the class).
"""
```

```python
"""
1. Magic method or Dunder Method:
-> Magic methods (dunder methods) in Python are special methods with double underscores like __init__ and __str__ that automatically define how objects behave in built-in operations.

2. Constructor:
-> it is a method which is define inside the class that's why it is called method. and it is called automatically when the object is created of the class that is it superpowers.

3. Why we use constructor?
-> The benefit of using a constructor is that it automatically initializes object data and setup when the object is created, making the code cleaner and more organized. Constructors solve the problem of manually initializing every object by automatically setting required values during object creation, which reduces repetitive code, prevents uninitialized objects, and makes development faster and safer.

4. Parameterized Constructor:
-> A parameterized constructor is a constructor that accepts arguments so objects can be initialized with different values during creation.
"""
```

```python
'''
1. __str__ method:
-> __str__ tells Python what text to show when you print an object.
-> When you print an object, Python automatically calls the __str__ method to get a string representation of the object.

2. __add__ method:
-> __add__ is a magic method that tells Python what should happen when you use the + operator between objects.

'''
```

```python
# Create our own data type Fraction:

class Fraction:
    # Parameterized Constructor
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return '{}/{}'.format(self.numerator, self.denominator)
    
    def __add__(self, other):
        new_numerator = self.numerator * other.denominator + self.denominator * other.numerator
        new_denominator = self.denominator * other.denominator
        return Fraction(new_numerator, new_denominator)
    
    def __sub__(self, other):
        new_numerator = self.numerator * other.denominator - self.denominator * other.numerator
        new_denominator = self.denominator * other.denominator
        return Fraction(new_numerator, new_denominator)

    def __mul__(self, other):
        new_numerator = self.numerator * other.numerator
        new_denominator = self.denominator * other.denominator
        return Fraction(new_numerator, new_denominator)
    
    def __truediv__(self, other):
        new_numerator = self.numerator * other.denominator
        new_denominator = self.denominator * other.numerator
        return Fraction(new_numerator, new_denominator)
    
    def convert_to_decimal(self):
        return self.numerator / self.denominator
```

```python
far = Fraction(3, 4)
dar = Fraction(1, 2)
print(far / dar)
```
