```python
# Inheritance:
# -> Inheritance in OOP means one class, called child class, can reuse the attributes and methods of another class, called parent class.

# Why we use Inheritance:
# -> We use inheritance to reuse common code from a parent class in child classes, so we avoid duplication and can extend or customize behavior easily.

# What actually get inherited:
# -> Non private Attributes and methods of the parent class
# -> Constructor of the parent class
```

```python
# parent_class
class User:

  def __init__(self):
    self.name = 'nitish'
    self.gender = 'male'

  def login(self):
    print('login')

# child_class
class Student(User):

  def __init__(self):
    self.rollno = 100

  def enroll(self):
    print('enroll into the course')
```

```python
u = User()
s = Student()
```

```python
s.login()
s.enroll()
# if we want to print(s.name) it will through an error because of method overwriting of parent class
```

```python
# constructor example 02:

class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self, price, brand, camera, os, ram):
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
        print ("Inside SmartPhone constructor")
```

```python

s = SmartPhone("Android", 2)
# if we want to access s.band() it will through an error because of method overwriting of parent class, the constructor of parent class is never executed so that the attributes or varible never created
```

```python
# child can't access private members of the class directly have to do wuth getter and setter:

# Parent_Class
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    # getter
    def show(self):
        print (self.__price)

# Child_Class
class SmartPhone(Phone):
    def check(self):
        print(self.__price)
```

```python
s=SmartPhone(20000, "Apple", 13)
s.show()
```

```python
# example 03:

# Parent_class
class Parent:

    def __init__(self,num):
        self.__num=num

    # getter
    def get_num(self):
        return self.__num

# Child_class
class Child(Parent):

    def show(self):
        print("This is in child class")
        
```

```python
son=Child(100)
print(son.get_num())
son.show()
```

```python
# example 04: if child has a constructor then it will override the parent constructor an the parent constructor will not be executed or called 

class Parent:

    def __init__(self,num):
        self.__num=num

    def get_num(self):
        return self.__num

class Child(Parent):

    def __init__(self,val,num):
        self.__val=val

    def get_val(self):
        return self.__val
```

```python
son = Child(100,10)
# print("Parent: Num:",son.get_num()), it will through an error because of method overwriting of parent class
print("Child: Val:",son.get_val())
```

```python
# Concept : Method Overriding
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
```
