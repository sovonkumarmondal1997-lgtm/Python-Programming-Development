```python
# find the top 3 genres by total earnings:
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genres with highest average IMDB rating:
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(3)
```

```python
# find the director with most poularity by votes:
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(3)
```

```python
# find the highest rated movies of each genre:
movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False).head(3)
```

```python
# find number of movies done by each actor:
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False).head(3)
```

```python
# agg method:
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list
genres.agg(['min','max','sum'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','max'],
        'IMDB_Rating':['min','max'],
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
df = pd.DataFrame(columns=movies.columns)

for group, data in genres:
    df = pd.concat(
        [df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]],
        ignore_index=True
    )

df
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal)
```

```python
# groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
```

```python
# find the best(in-terms of metascore(avg)) actor->genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(3)
```
