## GroupBy Methods:

### Aggregation Methods:

Method:

mean() : Mean of each group

sum() :	Sum of each group

min() :	Minimum value

max() :	Maximum value

count() : Count non-null values

size() : Count rows including nulls

std() :	Standard deviation

var() :	Variance

median() : Median

prod() : Product

sem() :	Standard error of mean

quantile() : Quantile values

describe() : Summary statistics

agg() / aggregate() : Apply multiple aggregations
