## Filtering Methods:

### Method:

filter() : Keep groups satisfying a condition

## Apply Methods:

### Method:

apply() : Apply a custom function

pipe() : Apply a function chain

## Missing Value Methods:

### Method:

fillna() : Fill missing values

ffill() : Forward fill

bfill() : Backward fill

## Utility Methods:

### Method:

sample() : Random sample from each group

value_counts() : Frequency counts within groups

corr() : Correlation by group

cov() :	Covariance by group

rolling() :	Rolling window operation

expanding() : Expanding window operation

resample() : Resampling within groups
