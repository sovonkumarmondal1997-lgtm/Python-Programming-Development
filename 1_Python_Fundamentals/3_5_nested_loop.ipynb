### Nested Loops:

- A nested loop is simply a loop inside another loop.

1. Row-Column Analogy: In coordinate tracking or grid problems, the outer loop tracking variable i traditionally represents the row number, while the inner loop tracking variable j represents the column number.

2. Execution Behavior: For every single iteration of the outer loop, the inner loop runs completely through its entire lifecycle from start to finish.

```python
# Nested Loop:
# -> Loop inside a Loop

# Unique pair pattern: [1-1,1-2,1-3,1-4,2-1,2-2,2-3,2-4,3-1,3-2,3-3,3-4,4-1,4-2,4-3,4-4]

for i in range(1,5):
    for j in range(1,5):
        print(i,j)

# In this loop 'i' represents the row number and 'j' represents the column number
```

```python
"""
Pattern Printing:
*
**
***
****
"""
# Take the row number form the user as input:
rows = int(input("Enter the no.of rows:"))
print(rows)

# logic for the pattern:
for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*',end='') # this print manages the '*' for a particular row
    print()  # this print manages the row 
```

```python
"""
Pattern Printing:
1
121
12321
1234321
"""
# Take the row number from the user as input:
rows = int(input("Enter the no.of rows: "))

for i in range(1, rows + 1):
    # Part 1: Climb Up (Starts at 1, ends at i)
    for j in range(1, i + 1):
        print(j, end='')
    
    # Part 2: Climb Down (Starts at i-1, ends at 1)
    # We use i-1 so we don't repeat the peak number
    for k in range(i - 1, 0, -1):
        print(k, end='')
    
    # Move to the next line after the row is complete
    print()
```

```python
"""
Loop Control Statement:
1. Break
2. Continue
3. Pass
"""
```

```python
# Break

for i in range(1,10):
    if i == 5:
        break
    print(i)
```

```python
# Identify the prime numbers between a range:

# Take the input from the users
lower_range = int(input("Enter the lower_range:"))
upper_range = int(input("Enter the upper_range:"))

for i in range(lower_range,upper_range+1):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
        print(i)
```

```python
# Continue:

for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
# Pass:

for i in range(1,10):
    pass
```
