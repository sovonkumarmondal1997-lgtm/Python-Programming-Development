```python
# Efficiency in programming is how well a program minimizes its use of execution time (time complexity) and computer memory (space complexity) to complete a task.

# Time efficiency: How fast your program runs, especially as the input gets bigger.

# Space efficiency: How much memory your program uses.

# We measure this with Big O notation, like O(1) for constant time, O(n) for linear time.

# Good efficiency means your program doesn't slow down too much or use too much memory when data grows.
```

```python
"""
Order of growth (Big O notation):
goals of order of growth:
- evaluate program's efficiency when input is very big
- wants to express the growth of program's run time as input size grows
- wants to put an upper bound on the growth of the program's run time as tight as possible
- do not need to be precise: order off not exact growth
- look at the largest factors in run time (which section of the program will take the longest to run)
Thus generally we want tight upper bound on the growth, as function of size of input, in worst case

"""
```

```python
# TIME COMPLEXITY - Simple Explanation
# =====================================

# Time complexity measures how long a program takes to run as the input size grows.
# Instead of measuring actual time (which depends on your computer), we count operations.

# Common Time Complexities (from fastest to slowest):
# O(1) - Constant: Always takes same time, no matter how much data. Example: accessing array[0]
# O(log n) - Logarithmic: Time grows slowly as data grows. Example: binary search
# O(n) - Linear: Time grows with the data size. Example: checking all items in a list
# O(n²) - Quadratic: Time grows very fast with data size. Example: nested loops
# O(2^n) - Exponential: Time doubles with each extra item. Very slow! Should avoid.
# O(n!) - Factorial: Extremely slow, rarely used.

# Real-world analogy:
# Finding a name in a phone book:
#   - O(1): If you know the page number (just look there)
#   - O(log n): If you split the book in half repeatedly (binary search)
#   - O(n): If you check every page one by one
#   - O(n²): If you compare every person with every other person
```

```python

# ============================================================
# DETAILED EXPLANATION: HOW TO MEASURE EACH TIME COMPLEXITY
# ============================================================

# 1. O(1) - CONSTANT TIME
# ========================
# Time is ALWAYS the same, no matter how big the input is.
# Even if you have 10 items or 10 million items, it takes same time.
# Example:
#   - Get first element: array[0]
#   - Access by key in dictionary: dict['name']
#   - Simple math operation: x = 5 + 3
# How to recognize: Code runs a fixed number of operations, no loops.

# 2. O(log n) - LOGARITHMIC TIME
# ================================
# Time grows VERY SLOWLY as data grows.
# If data doubles, time increases by just 1 step.
# Think of it like: Each step cuts data in half.
# Example:
#   - Binary search in sorted list
#   - Finding a person in phone book by splitting in half each time
#   - Tree height when splitting equally
# How to recognize: "Divide and conquer" approach, cutting data in half repeatedly.

# 3. O(n) - LINEAR TIME
# =======================
# Time grows DIRECTLY with data size.
# If you have 10 items, 10 operations. 100 items, 100 operations.
# Example:
#   - Finding max/min in unsorted list (check each item once)
#   - Counting how many items match a condition
#   - Simple for loop through all items
# How to recognize: Single for loop through all items, no nesting.

# 4. O(n²) - QUADRATIC TIME
# ===========================
# Time grows VERY FAST.
# If data doubles, time increases 4x!
# Example:
#   - Bubble sort, selection sort
#   - Comparing every item with every other item
#   - Nested loops: for i in list: for j in list: check(i, j)
# How to recognize: NESTED LOOPS (loop inside loop).

# 5. O(2^n) - EXPONENTIAL TIME
# ==============================
# Time DOUBLES with each new item added.
# With 10 items = 2^10 = 1,024 operations
# With 11 items = 2^11 = 2,048 operations (doubled!)
# Example:
#   - Generating all subsets of a set
#   - Fibonacci recursive without caching
#   - All combinations/permutations
# How to recognize: Functions that call themselves multiple times at each level.
# WARNING: Very slow! Should avoid if possible.

# 6. O(n!) - FACTORIAL TIME
# ===========================
# Time grows EXTREMELY FAST - worst complexity!
# n! = n × (n-1) × (n-2) × ... × 1
# With just 10 items = 3,628,800 operations
# Example:
#   - Generating all permutations of items
#   - Trying all possible orderings
# How to recognize: Multiple nested loops that generate permutations.
# WARNING: Never acceptable for large data. Avoid at all costs!

# ============================================================
# COMPARING SPEEDS (with 1000 items):
# ============================================================
# O(1)     = 1 operation
# O(log n) = ~10 operations
# O(n)     = 1,000 operations
# O(n²)    = 1,000,000 operations
# O(2^n)   = 2^1000 (impossible to calculate!)
# O(n!)    = 1000! (impossibly huge!)
```

```python
def fact_iter(n):
    answer = 1
    while n > 1:
        answer = answer * n
        n = n - 1
    return answer
```
