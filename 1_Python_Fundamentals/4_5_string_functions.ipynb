```python
# Replace: Replace any substring from the string 

var = 'I love my city Kolkata'
print(var)

var = 'I love my city Kolkata'.replace('Kolkata','Bengaluru')
print(var)
```

```python
# strip: remove whitespace from begining and ending not from middele 
name = '                 Kolkata love you'
print(name)

name = '                 Kolkata love you'.strip()
print(name)
```
