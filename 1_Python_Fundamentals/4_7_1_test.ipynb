```python
"""
Problem 1: Count Vowels
Take a string from the user and count how many vowels (a, e, i, o, u) are present.
"""
```

```python
text = input("Enter a string: ")
count = 0
for i in text:
    if(i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u'):
        count = count + 1
print("Number of vowels are: ", count)
```

```python
"""
Problem 2: Count Uppercase and Lowercase Letters

Take a string from the user and count:
- Number of uppercase letters
- Number of lowercase letters

Requirements:
- Use a loop.
- Do not use isupper() or islower().
- Compare characters manually using ranges:
'A' <= ch <= 'Z'
'a' <= ch <= 'z'
"""
```

```python
text = input("Enter a string: ")
count = 0
for i in text:
    if('A' <= i <= 'Z'):
        count = count + 1
print("Number of uppercase letters are: ", count)
count = 0
for i in text:
    if('a' <= i <= 'z'):
        count = count + 1
print("Number of lowercase letters are: ", count)
```

```python
"""
Problem 3 : Count Digits in a String

Take a string from the user and count how many digits are present.

Example:
Input: abc123xyz45
Output: 5

Requirements:
- Use a loop.
- Do not use isdigit().
- Check manually using:
'0' <= ch <= '9'
"""
```

```python
text = input("Enter a string: ")
count = 0
for i in text:
    if('0' <= i <= '9'):
        count = count + 1
print("Number of digits are: ", count)
```

```python
"""
Problem 4 : Count Spaces in a String
Take a string from the user and count how many spaces are present.

Example:
Input: I Love Python Programming
Output: 3

Requirements:
- Use a loop.
- Count only the space character ' '.
- Do not use count().
"""
```

```python
"""
Problem 5 : Count Special Characters
- Take a string from the user and count how many special characters are present.

Definition: A special character is anything that is not:
- an uppercase letter (A-Z)
- a lowercase letter (a-z)
- a digit (0-9)
- a space (' ')

Example:
Input: Hello@2026!
Output: 2
"""
```

```python
text = input("Enter a string: ")
count = 0
for i in text:
    if(not('A' <= i <= 'Z' or 'a' <= i <= 'z' or '0' <= i <= '9' or i == ' ')):
        count = count + 1
print("Number of special characters are: ", count)
```

```python
"""
Problem 6 : Reverse a String Without Using Slicing
- Take a string from the user and reverse it.

Example:
- Input: Python
- Output: nohtyP

Requirements:
- Do not use slicing:
"""
```

```python
"""
Problem 7 : Check Whether a String is a Palindrome
- A palindrome reads the same forward and backward.

Example:
- Input: madam
- Output: Palindrome

Example:
- Input: python
- Output: Not Palindrome

Requirements:
- Do not use:
    text[::-1]
- Use loops and string logic.
- Comparison should be case-insensitive.
"""
```

```python
text = input("Enter a string: ")
for i in range(len(text) - 1, -1, -1):
    if(text[i].lower() != text[len(text) - 1 - i].lower()):
        print("Not Palindrome")
        break
else:
    print("Palindrome")
```

```python
"""
Problem 8 : Count the Frequency of a Character

Take:
- A string
- A character
- Count how many times that character appears in the string.

Example:
Input: banana
- a
Output: 3

Requirements:
- Use a loop.
- Do not use count().
- Comparison should be case-insensitive.
"""
```

```python
text = input("Enter a string: ")
char = input("Enter a character: ")
count = 0
for i in text:
    if(i.lower() == char.lower()):
        count = count + 1
print("Frequency of", char, "is", count)
```

```python
"""
Problem 11 (Hard)

Find the First Non-Repeating Character

Take a string from the user and print the first character that appears exactly once.

Example
Input: swiss

Output: w

Explanation:

s → appears 3 times
w → appears 1 time ← first non-repeating
i → appears 1 time
Example
Input: aabbccde

Output: d

Input: aabbcc

Output: No non-repeating character

Requirements
Do not use dictionaries.
Do not use count().
Use loops only.
Comparison should be case-insensitive.
"""
```

```python
text = input("Enter a string: ")
for i in text:
    count = 0
    for j in text:
        if(i.lower() == j.lower()):
            count = count + 1
    if(count == 1):
        print("First non-repeating character is: ", i)
        break
else:
    print("No non-repeating character")
```

```python
"""
Problem 12 (Hard)

Find the Character with Maximum Frequency

Take a string from the user and print:

The character that appears the most
Its frequency
Example
Input: banana

Output:
a 3
Example
Input: Mississippi

Output:
i 4

(Comparison should be case-insensitive.)

Requirements
Do not use dictionaries.
Do not use count().
Use loops only.
If multiple characters have the same highest frequency, print the one that appears first in the string.
Example
Input: aabbbcc
Output:
b 3
"""
```

```python
text = input("Enter a string: ")
max_char = ''
max_count = 0
for i in text:
    count = 0
    for j in text:
        if(i.lower() == j.lower()):
            count = count + 1
    if(count > max_count):
        max_char = i
        max_count = count
print("Character with maximum frequency is: ", max_char)
print("Frequency of", max_char, "is: ", max_count)
```

```python
"""
Problem 13 (Hard)

Remove Duplicate Characters

Take a string from the user and create a new string containing only the first occurrence of each character.

Example
Input: banana

Output: ban

Explanation:

b → keep
a → keep
n → keep
a → already seen
n → already seen
a → already seen
Example
Input: Mississippi

Output: Misp

(Comparison should be case-insensitive.)

Requirements
Do not use sets.
Do not use dictionaries.
Use loops only.
Preserve the original order.
Keep the first occurrence.
"""
```

```python
text = input("Enter a string: ")
new_text = ''
for i in text:
    if i not in new_text:
        new_text = new_text + i
print("New string is: ", new_text)
```

```python
"""
Problem 14 (Hard)

Find the Longest Word

Take a sentence from the user and print the longest word.

Example
Input:
I love Python programming

Output:
programming
Example
Input:
Data Engineering is awesome

Output:
Engineering
Requirements
Do not use split().
Use loops only.
Assume words are separated by a single space.
If multiple words have the same maximum length, print the first one.
"""
```

```python
text = input("Enter a string: ")

current_word = ""
max_word = ""

for ch in text:
    if ch != ' ':
        current_word = current_word + ch
    else:
        if len(current_word) > len(max_word):
            max_word = current_word

        current_word = ""

if len(current_word) > len(max_word):
    max_word = current_word

print("Longest word is:", max_word)
```

```python
"""
Problem 15 (Hard)

Find the Most Frequent Word

Take a sentence from the user and print:

The word with the highest frequency
Its frequency
Example
Input:
apple banana apple mango apple banana

Output:
apple 3
Example
Input:
python data python engineering data python

Output:
python 3
Requirements
Do not use dictionaries.
Do not use count().
Comparison should be case-insensitive.
If multiple words have the same frequency, print the one that appears first.
You may use split() for this problem.
"""
```

```python
text = input("Enter a string: ")
words = text.split()
max_word = ''
max_count = 0
for i in words:
    count = 0
    for j in words:
        if(i.lower() == j.lower()):
            count = count + 1
    if(count > max_count):
        max_word = i
        max_count = count
print("Word with maximum frequency is: ", max_word)
print("Frequency of", max_word, "is: ", max_count)
```
