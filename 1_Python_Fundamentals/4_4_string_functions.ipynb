```python
"""
1. Count 
2. Find 
3. Index
"""
```

```python
# count: it count how many times a paricular string(character) comes or present in the string
'Hi, currently i am learning python'.count('e')
```

```python
# find: find the index position of a character or a substring
'Hi, currently i am learning python'.find('e')
```

```python
# find: find the index position of a character or a substring
'Hi, currently i am learning python'.find('am')
```

```python
# find: find the index position of a character or a substring
'Hi, currenlt i am learning python'.find('yy')

# when a substring or character is not present in the string it's shows: -1
```

```python
# index: it's also find the index position of a character or a substring
'Hi, currenlt i am learning python'.index('python')

# when a substring or character is not present in the string it's shows: Error

# this is the difference between index and find
# find shows: -1 and index shows: Error [for missing value]
```
