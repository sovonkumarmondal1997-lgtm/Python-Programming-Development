```python
"""
1. Prints the Simple Interest.
Use input().
Convert inputs to the appropriate numeric type.
Do not use loops, functions, or conditionals yet.
"""
```

```python
principal = float(input("Enter the principal amount: "))
print(f"The principal amount is: {principal}")

rate_of_interest = float(input("Enter the rate of interest: "))
print(f"The rate of interest is: {rate_of_interest}%")

time_period = float(input("Enter the time period: "))
print(f"The time period is: {time_period}")

simple_interest = (principal * rate_of_interest * time_period) / 100

print("The simple interest is: ", simple_interest)
```

```python
"""
Problem 3: Temperature Converter (Celsius to Fahrenheit)

Write a Python program that:
a. Takes temperature in Celsius as input.
b. Converts it to Fahrenheit using:
"""
```

```python
celsius = float(input("Enter the temperature in Celsius: "))

fahrenheit = (celsius * 1.8) + 32

print(f"The temperature in Fahrenheit is: {fahrenheit}")
```

```python
"""
Problem 4: Total Marks and Percentage

Write a Python program that:

a. Takes marks of 5 subjects as input.
b. Calculates the total marks.
c. Calculates the percentage.
"""
```

```python
subject1 = int(input("Enter the marks of subject 1: "))
subject2 = int(input("Enter the marks of subject 2: "))
subject3 = int(input("Enter the marks of subject 3: "))
subject4 = int(input("Enter the marks of subject 4: "))
subject5 = int(input("Enter the marks of subject 5: "))

total_marks = subject1 + subject2 + subject3 + subject4 + subject5

percentage = (total_marks / 500) * 100

print(f"The total marks is: {total_marks}")
print(f"The percentage is: {percentage}%")
```

```python
"""
Problem 5: Swap Two Numbers (Without Using a Third Variable)

Write a Python program that:

a. Takes two numbers as input.
b. Swaps their values without using a third variable.
c. Prints the values before and after swapping.
"""
```

```python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

print(f"Before swapping:")
print(f"The first number is: {a}")
print(f"The second number is: {b}")

a, b = b, a

print(f"After swapping:")
print(f"The first number is: {a}")
print(f"The second number is: {b}")
```

```python
"""
Problem 6: Convert Total Days into Years, Months, and Days

Write a Python program that:

Takes the total number of days as input.

Converts them into:
a. Years (1 year = 365 days)
b. Months (1 month = 30 days)
c. Remaining Days
"""
```

```python
total = int(input("Enter the total number of days: "))

years = total // 365
months = (total % 365) // 30
days = (total % 365) % 30

print(f"The total number of years is: {years}")
print(f"The total number of months is: {months}")
print(f"The total number of days is: {days}")
```

```python
"""
Problem 7: Calculate the Average of 3 Numbers

Write a Python program that:

a. Takes 3 numbers as input.
b. Calculates their sum.
c. Calculates their average.
"""
```

```python
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))
number3 = int(input("Enter the third number: "))

sum = number1 + number2 + number3

average = sum / 3

print(f"The sum is: {sum}")
print(f"The average is: {average}")
```

```python
"""
Problem 8: Convert Seconds into Hours, Minutes, and Seconds

Write a Python program that:

i. Takes total seconds as input.
ii. Converts them into:
    a. Hours
    b. Minutes
    c. Remaining Seconds
"""
```

```python
total_seconds = int(input("Enter the total number of seconds: "))

hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = (total_seconds % 3600) % 60

print(f"The total number of hours is: {hours}")
print(f"The total number of minutes is: {minutes}")
print(f"The total number of seconds is: {seconds}")
```

```python
"""
Problem 9: Profit or Loss Calculator

Write a Python program that:

i. Takes the Cost Price (CP) as input.
ii. Takes the Selling Price (SP) as input.

Calculates:
    a. Profit = SP - CP
    b. Loss = CP - SP
"""
```

```python
cost_price = float(input("Enter the cost price: "))
selling_price = float(input("Enter the selling price: "))

profit = selling_price - cost_price
loss = cost_price - selling_price

print(f"The profit is: {profit}")
print(f"The loss is: {loss}")
```

```python
"""
Problem 10: Total Amount in Rupees

Write a Python program that takes:

Number of ₹2000 notes as input
Number of ₹500 notes as input
Number of ₹200 notes as input
Number of ₹100 notes as input

and calculates the total amount of money.
"""
```

```python
two_thousands = int(input("Enter the number of ₹2000 notes: "))
five_hundreds = int(input("Enter the number of ₹500 notes: "))
two_hundreds = int(input("Enter the number of ₹200 notes: "))
one_hundreds = int(input("Enter the number of ₹100 notes: "))

total_amount = (two_thousands * 2000) + (five_hundreds * 500) + (two_hundreds * 200) + (one_hundreds * 100)

print(f"The total amount is: {total_amount}")
```
