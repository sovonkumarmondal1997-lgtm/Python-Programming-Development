```python
# in output between every data type there is a 'space'. it is happening for a default print setting.
# which we can see in the manual of print module. that is sep=''
print('Hello',1,7.898,True)
```

```python
help(print)
```

```python
# here we change the default setting from sep=' ' to sep='/' only for this reason when we print, all the datatypes together we can see '/' between them 

# using 'sep' keword we can use anything to print the desired result

print('Hello',1,7.898,True,sep='/')
```

```python
# we can see different print functions show the result as output in different line not in a single line

# in print function there is a default option is end='\n', \n = line change, that is the reason,
# when we use print module in different line it shows output in different line not in a single line

print("Hello")
print("World")
```

```python
help(print)
```

```python
# here we overwrite the default value '\n' with '-' thats why it is not changing the line,
# and showing output in one line with '-'
print("Hello",end='-')
print("World")
```

```python
print("Hello",end=' ')
print("World")
```

Data Types

```python
# 1. Integer:
print(7)

# total integer limit in python:
print(10**308)   # meaining: 1*10^308
```

```python
# 2. Decimal/FLoat:
print(11.88)

# total float limit in python:
print(1.7e308)  # meaning: 1.7*10^308
```

```python
# 3. Boolean:
print(True)
print(False)
```

```python
# 4. text/string:
print("Hello World")
print('Hello World')
```
