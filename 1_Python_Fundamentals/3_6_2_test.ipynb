```python
"""
Problem 1: Sum of First N Natural Numbers
Write a program that takes a number n from the user and calculates the sum of every digit of the number.
1+2+3+4+5+...n

Requirements:
- Use a for loop.
- Do not use the built-in sum() function.
- Use an accumulator variable.
"""
```

```python
number = int(input("Enter a number: "))
total = 0

# take every digit of the number
for i in range(1, number + 1):
    total += i

print(f"The sum of the first {number} natural numbers is: {total}")
```

```python
"""
Problem 2: Sum of Squares
Write a program that takes a number n and calculates:
1² + 2² + 3² + ... + n²

Requirements:
- Use a for loop.
- Use an accumulator variable.
- Do not use any mathematical shortcut formula.
"""
```

```python
number = int(input("Enter a number: "))
total = 0

for i in range(1, number + 1):
    i = i ** 2
    total += i

print(f"The sum of the first {number} natural numbers is: {total}")
```

```python
"""
Problem 3: Multiplication Table
Write a program that takes a number from the user and prints its multiplication table up to 10.
7 * 1 = 7
7 * 2 = 14
7 * 3 = 21
...
7 * 10 = 70

Requirements:
- Use a for loop.
- Print the output exactly in the format shown above.
- Do not use a nested loop.
"""
```

```python
number = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{number} * {i} = {number * i}")
```

```python
"""
Problem 4: Factorial of a Number
Write a program that takes a number n and calculates: n!
5! = 5 * 4 * 3 * 2 * 1 = 120

Requirements:
- Use a for loop.
- Do not use math.factorial().
- Use an accumulator variable.
- Solve it manually
"""
```

```python
number = int(input("Enter a number: "))
total = 1

for i in range(1, number + 1):
    total *= i

print(f"The factorial of {number} is: {total}")
```

```python
"""
Problem 5: Count Multiples of 3
Write a program that takes a number n and counts how many numbers between 1 and n are divisible by 3.

Requirements:
- Use a for loop.
- Use an if statement.
- Maintain a counter variable.
- Do not use any mathematical shortcut formula.
"""
```

```python
number = int(input("Enter a number: "))
count = 0

for i in range(1, number + 1):
    if i % 3 == 0:
        count += 1

print(f"There are {count} numbers between 1 and {number} that are divisible by 3.")
```

```python
"""
Problem 6: Print All Factors of a Number
Write a program that takes a number n and prints all its factors.

Example:

Input: 12

Output:

1
2
3
4
6
12
"""
```

```python
"""
Problem 7: Count Factors of a Number
Write a program that takes a number n and counts how many factors it has.

Example:

Input: 12
Factors: 1, 2, 3, 4, 6, 12
Output: 6
"""
```

```python
number = int(input("Enter a number: "))
count = 0

for i in range(1, number + 1):
    if number % i == 0:
        count += 1

print(f"There are {count} factors of {number}.")
```

```python
"""
Problem 8: Check Whether a Number is Prime
A number is prime if it has exactly 2 factors: 1 and itself
"""
```

```python
number = int(input("Enter a number: "))
count = 0

for i in range(1, number + 1):
    if number % i == 0:
        count += 1

if count == 2:
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
"""
Problem 9: Sum of All Factors of a Number
Write a program that takes a number n and finds the sum of all its factors.

Example:

Input: 12
Factors: 1, 2, 3, 4, 6, 12
Sum: 1 + 2 + 3 + 4 + 6 + 12 = 28
Output: 28
"""
```

```python
number = int(input("Enter a number: "))
sum = 0

for i in range(1, number + 1):
    if number % i == 0:
        sum += i

print(f"The sum of all factors of {number} is: {sum}")
```

```python
"""
Problem 10: Print a Right-Angled Star Triangle
Write a program that takes the number of rows from the user and prints:

Example

Input: 5

Output:

*
**
***
****
*****

Requirements:
- Use nested for loops.
- Do not use string multiplication like:
"""
```

```python
number = int(input("Enter the number of rows: "))

# This is the outer loop, It controls how many rows will be printed.
# it starts from 1 and go to the number entered by the user
for i in range(1, number + 1):

    # This is the inner loop.It controls how many stars (*) are printed in the current row.
    # The number of stars depends on the current value of i.
    for j in range(1, i + 1):

        # Prints one *.end="" tells Python not to move to the next line after printing.
        # The next star is printed on the same line.
        print("*", end="")

        # Prints a blank line (moves the cursor to the next line).
        # It runs after the inner loop finishes.This starts printing stars for the next row.
    print()
```
