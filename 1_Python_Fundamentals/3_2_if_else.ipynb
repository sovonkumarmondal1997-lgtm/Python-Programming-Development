```python
# Writing if/elif/else statement for a website login: (basic understanding of if/else statement)
# assume login credential is: email = sovon@campusx.in, password = 1234

# Take the input from the user:
email = input("Please Enter Your Email:")
password = input("Please Enter Your Password:")

# Checked if the entered credentials match the expected values:
if email == 'sovon@campusx.in' and password == '1234':
    print("You are Welcome")
elif email == 'sovon@campusx.in' and password != '1234':
    print("Incorrect Password")
    password = input("Please Enter Your Password:")
    if password == '1234':
        print("Finally Welcome!")
    else:
        print("Again Incorrect Password! Try After Sometimes")
else:
    print("Please Try Again!")
```

```python
# Find the minimum number between three numbers:

# Take the input from the users:
a = int(input("Enter the first number (a):"))
b = int(input("Enter the second number (b):"))
c = int(input("Enter the third number (c):"))

# Checke the possibilites to indentify the minimum number:
if a < b and a < c:
    print("Smallest number is a:",a)
elif b < c:
    print("Smallest number is b:",b)
else:
    print("Smallest numbr is c:",c)
```

```python
# program a simple menu driven calculator:

# Take the input from the user:
number_1 = int(input("Enter the first number:"))
number_2 = int(input("Enter the second number:"))
operator = input("Enter the operation value ('+','-','*','/'): ")

# Check the possibilities as per the input:
if operator == '+':
    addition = number_1 + number_2
    print("Sum of two number is:",addition)
elif operator == '-':
    substraction = number_1 - number_2
    print("Substraction of two number is:",substraction)
elif operator == '*':
    multiplication = number_1 * number_2
    print("Multiplication of two number is:",multiplication)
else:
    division = number_1 / number_2
    print("Division of two number is:",division)
```
