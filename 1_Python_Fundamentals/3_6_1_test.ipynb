```python
"""
Problem 1: Count Even and Odd Numbers

Write a program that takes a number n from the user and counts:
- How many even numbers are between 1 and n
- How many odd numbers are between 1 and n

Requirements
- Use a loop.
- Use an if-else condition.
- Do not use any built-in shortcuts.
"""
```

```python
number = int(input("Enter the number: "))

odd = 0
even = 0

for i in range(1, number+1):
    if i % 2 == 0:
        even = even + 1
    else:
        odd = odd + 1

print(f"There are {even} even numbers and {odd} odd numbers between 1 and {number}")
```

```python
"""
Problem 2: Reverse a Number
- Write a program that takes a number from the user and prints the number in reverse order.

Requirements:
- Use a while loop.
- Use % and //.
- Do not convert the number to a string.
- Build the reversed number mathematically.
"""
```

```python
number = int(input("Enter the number: "))

reversed_number = 0

while number > 0:
    digit = number % 10
    reversed_number = reversed_number * 10 + digit
    number = number // 10

print(f"The reversed number is: {reversed_number}")
```

```python
"""
Problem 3: Sum of Digits
- Write a program that takes a number from the user and calculates the sum of all its digits.

Requirements
- Use a while loop.
- Use % and //.
- Do not convert the number to a string.
- Build the logic mathematically.
"""
```

```python
number = int(input("Enter the number: "))
print(f"The number entered by the user is: {number}")

sum = 0

while number > 0:
    digit = number % 10
    sum = sum + digit
    number = number // 10

print(f"The sum of the digits is: {sum}")
```

```python
"""
Problem 4: Count the Number of Digits
- Write a program that takes a positive integer and counts how many digits it contains.

Requirements
- Use a while loop.
- Use //.
- Do not convert the number to a string.
- Count the digits mathematically.
"""
```

```python
number = int(input("Enter the number: "))
print(f"The number entered by the user is: {number}")

count = 0

while number > 0:
    number = number // 10
    count = count + 1

print(f"The number of digits in the number is: {count}")
```

```python
"""
Problem 5: Find the Largest Digit in a Number
- Write a program that takes a number from the user and finds the largest digit present in that number.

Requirements:
- Use a while loop.
- Use % and //.
- Do not convert the number to a string.
- Compare digits mathematically.
"""
```

```python
number = int(input("Enter the number: "))
print(f"The number entered by the user is: {number}")

largest_digit = 0

while number > 0:
    digit = number % 10
    if digit > largest_digit:
        largest_digit = digit
    number = number // 10

print(f"The largest digit in the number is: {largest_digit}")
```

```python
"""
Problem 6: Check Whether a Number is a Palindrome
- A palindrome number reads the same forward and backward.

Requirements
Use a while loop.
Use % and //.
Do not convert the number to a string.
Compare the original number with the reversed number.
"""
```

```python
number = int(input("Enter the number: "))
print(f"The number entered by the user is: {number}")


original_number = number
reversed_number = 0

while number > 0:
    digit = number % 10
    reversed_number = reversed_number * 10 + digit
    number = number // 10

if reversed_number == original_number:
    print(f"The number {original_number} is a palindrome.")
else:
    print(f"The number {original_number} is not a palindrome.")
```

```python
"""
Problem 7: Product of Digits
- Write a program that takes a number from the user and calculates the product of all its digits.

Requirements
- Use a while loop.
- Use % and //.
- Do not convert the number to a string.
- Solve it mathematically.
"""
```

```python
"""
Problem 9: Check Whether a Number is an Armstrong Number
- A 3-digit Armstrong number is a number where:
    (sum of cubes of its digits) = original number

Requirements:
- Use a while loop.
- Use % and //.
- Store the original number before modifying it.
- Assume the input is always a 3-digit number.
"""
```

```python
number = int(input("Enter the number: "))
print(f"The number entered by the user is: {number}")

original_number = number

sum = 0

while number > 0:
    digit = number % 10
    sum = sum + digit ** 3
    number = number // 10

if sum == original_number:
    print(f"The number {original_number} is an Armstrong number.")
else:
    print(f"The number {original_number} is not an Armstrong number.")
```

```python
"""
Problem 10: Print All Prime Numbers in a Range
- Write a program that takes two numbers from the user:

Start Range
End Range

and prints all prime numbers between them.

Requirements:
- Use nested loops.
- Use break.
- Use the for-else concept you learned.
- Do not use any built-in prime-checking functions
"""
```

```python
start_range = int(input("Enter the start range: "))
end_range = int(input("Enter the end range: "))

for i in range(start_range, end_range+1):
    if i < 2:
        continue

    for j in range(2, i):
        if i % j == 0:
            break
    else:
        print(i)
```
