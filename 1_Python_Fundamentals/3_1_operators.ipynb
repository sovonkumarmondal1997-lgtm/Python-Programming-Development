Operators in Python:
1. Arithmetic operators
2. Relational Operators
3. Logical Operators
4. Bitwise Operators
5. Assignment Operators
6. Membership Operators

```python
# 1. Arithmetic Operators:

# Addition 
print(5+5)

# Substruction
print(10-5)

# Multiplication
print(5*5)

# Division
print(10/2)

# Integer-Division or floor-division
print(5//2)

# Modulas 
print(5%2)

# Power-off
print(5**2)
```

```python
# Relationl Operators:

print(4 > 5)
print(4 < 5)
print(4 >= 5)
print(4 <= 5)
print(4 == 5)
print(4 != 5)
```

```python
# Logical operators:

print(1 and 0)
print(1 or 0)
print(not 1)
```

```python
# Bitwise Operators: It works on binary value

# Bitwise_and
print(2 & 3)

# Bitwise_or
print(2 | 3)

# Bitwise_xor
print(2 ^ 3)

# Bitwise_not
print(~3)

# Bitwise_leftshift
print(4 << 1)

# Bitwise_rightshift
print(4 >> 1)
```

```python
# Find the sum of a three digit number entered by the users:

# Take the input from the user:
N = int(input("Enter the three-digit number: "))

# Extracting single numbers from the three digit:
a = N // 100
N = N % 100
b = N // 10
N = N % 10
c = N % 10

# Calculate the sum of the three digit:
total_sum = (a+b+c)

# Display the result:
print("Total sum of the three-digit number is:",total_sum)
```

```python
# Find the sum of a three digit number entered by the users:

# Take the input from the user:
N = int(input("Enter the three-digit number: "))

# Extracting single numbers from the three digit:
hundreds = N // 100
tens = (N % 100) // 10
units = (N % 100) % 10

# Calculate the sum of the three digit:
total_sum = (hundreds + tens + units)

# Display the result:
print("Total sum of the three-digit number is:",total_sum)
```
