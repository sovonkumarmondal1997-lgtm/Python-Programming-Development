```python
# Ceating a class:
class Atm:

    #constructor (special function):
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu() # executing the menu inside the constructor

    # Create a functaionality so user can see a menu of what they can do
    def menu(self):
        user_input = input("""
        Hello, how can I help you?
        1. Press 1 to create a pin
        2. Press 2 to change a pin
        3. Press 3 to check your balance
        4. Press 4 to withdraw
        5. Press 5 to exit
        """)

        if user_input == "1":
            # creating a pin
            self.create_pin()
        elif user_input == "2":
            # changing a pin
            self.change_pin()
        elif user_input == "3":
            # check balanc
            self.check_balance()
        elif user_input == "4":
            # withdraw
            self.withdraw()
        elif user_input == "5":
            # exit
            print("Thanks for using our service")
            return
        else:
            print("You have selected a wrong option")
            self.menu()

    def create_pin(self):
        user_pin = input("Enter your pin: ")
        self.pin = user_pin

        user_balance = int(input("Enter your balance: "))
        self.balance = user_balance
        print("Pin has been created successfully")
        self.menu()

    def change_pin(self):
        old_pin = input("Enter your old pin: ")
        if old_pin == self.pin:
            # Let the user to change the pin
            new_pin = input("Enter your new pin: ") 
            self.pin = new_pin
            print("Pin has been changed successfully")
            self.menu()
        else:
            print("You have entered wrong existing pin")
            self.menu()
    
    def check_balance(self):
        user_pin = int(input("Enter your pin: "))
        if user_pin == self.pin:
            print(f"Your balance is {self.balance}")
        else:
            print("You have entered wrong pin")
            self.menu()

    def withdraw(self):
        user_pin = input("Enter your pin: ")
        if user_pin == self.pin:
            amount = int(input("Enter the amount: "))
            if amount <= self.balance:
                self.balance -= amount
                print("Withdrawal successful")
                print(f"Current balance is {self.balance}")
            else:
                print("Insufficient balance")
                print(f"Current balance is {self.balance}")
        else:
            print("You have entered wrong pin")
            self.menu()
```
