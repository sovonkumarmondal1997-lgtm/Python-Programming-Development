```python
# Class Relationship:
# 1. Aggregation (One class uses/has another class object, but both can exist independently.)
# 2. Inheritance (One class gets properties and methods from another class.)
```

```python
# Aggregation(Has-A relationship):
class Customer:

  def __init__(self,name,gender,address):
    self.name = name
    self.gender = gender
    self.address = address

  def print_address(self):
    print(self.address.get_city(),self.address.pin,self.address.state)

class Address:

  def __init__(self,city,pin,state):
      self.__city = city
      self.pin = pin
      self.state = state

  def get_city(self):
    return self.__city
  
# we can not access private attributes or variable when we performing aggregation relationship because in aggregation one class only holds another class’s object reference, but private attributes are protected by that object’s own class using name-mangling/encapsulation.
# in aggregation, we should access or update another object’s private attributes using its public getter and setter methods
```

```python
address1 = Address('Gariahat',700006,'Kolkata')
customer1 = Customer('Rahul Kumar','male',address1)
customer1.print_address()
```

```python
add1 = Address('gurgaon',122011,'haryana')
cust = Customer('nitish','male',add1)

cust.print_address()

cust.edit_profile('ankit','mumbai',111111,'maharastra')
cust.print_address()
```
