```python
"""
OOP = Object Oriented Programming Principals:
1. Class
2. Object
3. Polymorphism
4. Encapsulation
5. Inheritance
6. Abstraction

In Python, list, dict, and str are classes because they define behavior and structure, and every actual value like [1,2], {"a":1}, or "hello" is an object (instance) created from those classes.

1. What is a class?
-> A class in Python OOP is a blueprint/template that defines the data (attributes) and behavior (methods) for creating objects.

-> Inside an OOP class, we define attributes (variables/data) and methods (functions/behavior) that describe the objects created from that class.

2. What is an object?
-> An object in Python is an instance of a class that contains real data and can use the methods defined in that class.
"""
```

```python
"""
1. Creating a class name: 

Pascal Case (also called Upper Camel Case) is a naming style where every word starts with a capital letter and no spaces or underscores are used.

Example:
StudentData
BankAccount
EmployeeRecord

2. COnstructor:
A constructor in OOP is a special method that automatically runs when an object is created to initialize the object's data.

In Python, the constructor method is __init__().

3. If we want to make the variable inside the class:
-> We have to make it inside the constructor
-> We have to use self.variable_name
"""
```

```python
# Ceating a class:
class Atm:

    #constructor
    def __init__(self):
        self.pin = ""
        self.balance = 0
```

```python
# Making the object of the class: object_name = classneme()
obj = Atm()
```

```python
# check the type of the class:
print(type(obj))
```

```python
# When you write obj., we are trying to access the attributes (variables) and methods (functions) that belong to that object using the dot operator (.).
obj.balance
obj.pin
```
