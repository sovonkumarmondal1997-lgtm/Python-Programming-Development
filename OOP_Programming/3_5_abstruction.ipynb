```python
# Abstruction:
# -> Abstraction in OOP means hiding complex internal details and showing only the necessary features to the user.

# Why should we use Abstruction?
# -> We use abstraction to hide complex implementation details and expose only what is necessary, making code easier to use, maintain, and change.

# Example: when using car.start(), you don’t need to know the internal engine process—you only use the simple interface.
```

```python
# Abstruct class:
# -> An abstract class is an incomplete class that is used as a plan for other classes.
# -> You cannot create an object from it directly.

# how to create abstruct class?
from abc import ABC, abstractmethod

# what is abstract method?
# -> An abstract method is a method declared in an abstract class that has no complete code and must be implemented by child classes.

# Why we use abstruct class and abstruct method?
# -> We use abstract class and abstract method to create a common rule/blueprint that child classes must follow.
```

```python
from abc import ABC,abstractmethod
class BankApp(ABC):

  def database(self):
    print('connected to database')

  @abstractmethod
  def security(self):
    pass

  @abstractmethod
  def display(self):
    pass
```

```python
class MobileApp(BankApp):

  def mobile_login(self):
    print('login into mobile')

  def security(self):
    print('mobile security')

  def display(self):
    print('display')
```

```python
mob = MobileApp()
```

```python
mob.security()
```
