```python
"""
Reference Variables:
Reference variables hold the objects
We can create objects without reference variable as well
An object can have multiple reference variables
Assigning a new reference variable to an existing object does not create a new object
"""
```

```python
# object without a reference:
class Person:

  def __init__(self):
    self.name = 'nitish'
    self.gender = 'male'

Person()
# here the oject is created without a reference variable, here the variable name is not the object itself, it has a memory adrees reference to the object
```

```python
# multiple reference variable:
p = Person()
q = p
print(p)
print(q)
# here the two reference variables point to the same object
print(id(p))
print(id(q))
```

```python
# change attribute value with the help of 2nd object:
print(p.name)
print(q.name)

q.name = 'sovon kumar'
print(p.name)
print(q.name)
```

```python
# Pass by reference:
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
  print('Hi my name is',person.name,'and I am a',person.gender)
  p1 = Person('ankit','male')
  return p1

p = Person('nitish','male')
# here the object p is passed by reference
greet(p)

# here the object p is passed by reference
x = greet(p)
print(x.name)
print(x.gender)
```

```python
# mutabiliy of the object:
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
    # 'person' receives the same object reference as 'p'
    # It does NOT create a new Person object

    # Since objects are mutable, we can change their attributes
  person.name = 'ankit'
  # Returning the same modified object
  return person

# Creating one Person object
p = Person('nitish','male')

# id(p) shows the memory identity/reference of object p
print(id(p))

# Passing object p to greet()
# Inside greet(), 'person' points to the same object as p
p1 = greet(p)

# p1 receives the same object returned from greet()
# So p and p1 both point to the same object
print(id(p1))

# Both id(p) and id(p1) are same, but the object’s name changed from "nitish" to "ankit", proving the same object was modified instead of creating a new object.
```
