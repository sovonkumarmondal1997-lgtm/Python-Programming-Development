```python
# Polymorphism:
# -> Polymorphism in OOP means the same method name can behave differently depending on the object that calls it

# 1. Method Overriding:
# -> A child class gives its own version of a method already defined in the parent class.

# 2. Method Overloading:
# -> Same method name is used with different parameters, but Python does not support traditional method overloading directly.

# 3. Operator Overloading:
# -> Giving special meaning to operators like +, -, or * for user-defined objects/classes.
```

```python
# Method Overriding means a child class changes the behavior of a method that already exists in the parent class.
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):   # overriding parent method
        print("Dog barks")
```

```python
animal = Animal()
dog = Dog()

animal.sound()  # Animal makes a sound
dog.sound()     # Dog barks
```

```python
# Why should we use method overriding?
# -> We use method overriding because it allows a child class to reuse the parent class structure but customize specific behavior according to its own need.
```

```python
# Method Overloading means using the same method name with different numbers or types of parameters.
#In Python, traditional method overloading is not directly supported, but we can achieve it using default arguments.

class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c
```

```python
calc = Calculator()

print(calc.add(5))        # 5
print(calc.add(5, 10))    # 15
print(calc.add(5, 10, 20)) # 35

# Here, the same method add() works differently based on how many arguments we pass, so it behaves like method overloading.
```

```python
# We use method overloading because it lets one method handle different numbers or types of inputs, making code cleaner and easier to use.
```

```python
# Operator Overloading means giving an operator like +, -, or * a custom meaning for objects.
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks
```

```python
s1 = Student(80)
s2 = Student(90)

print(s1 + s2)  # 170

# Here, + normally adds numbers, but we overloaded it using __add__() so it can add the marks of two Student objects.
```

```python
# why should we use operator overloading?
# -> We use operator overloading to make custom objects behave naturally with operators like +, -, *, etc., which makes code more readable and intuitive.
```
