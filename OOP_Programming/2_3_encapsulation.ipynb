```python
# Getters and Setters:
# -> Getters and setters are public methods used to safely read and update private instance variables in a class.

class BankAccount:

    def __init__(self, balance):
        self.__balance = balance   # private instance variable

    # Getter method: used to read/get the private variable value
    def get_balance(self):
        return self.__balance

    # Setter method: used to update/set the private variable value
    def set_balance(self, balance):
        if balance >= 0:
            self.__balance = balance
        else:
            print("Balance cannot be negative")


account = BankAccount(10000)

# Calling getter
print(account.get_balance())   # Output: 10000

# Calling setter
account.set_balance(15000)

# Calling getter again after update
print(account.get_balance())   # Output: 15000
```

```python
class Atm:

  __counter = 1
  # Static variable / Class variable
  # Shared by all Atm objects
  # Used here to generate unique customer id/cid

  def __init__(self):
    print(id(self))

    self.pin = ''
    # Instance variable
    # Separate pin for each Atm object

    self.__balance = 0
    # Private instance variable
    # Separate balance for each Atm object

    self.cid = Atm.__counter
    # Instance variable
    # Separate cid for each Atm object

    Atm.__counter = Atm.__counter + 1
    # Updating static/class variable
    # Shared counter increases whenever a new object is created

  @staticmethod
  def get_counter():
    return Atm.__counter
    # Accessing static/class variable
```

```python
class Student:
    school_name = "ABC School"   # static/class variable shared by all objects

    def __init__(self, name):
        self.name = name         # instance variable unique for each object


s1 = Student("Rahul")
s2 = Student("Anita")

print(s1.name)           # Rahul
print(s2.name)           # Anita

print(s1.school_name)    # ABC School
print(s2.school_name)    # ABC School

Student.school_name = "XYZ School"  # changing static variable for all objects

print(s1.school_name)    # XYZ School
print(s2.school_name)    # XYZ School
```
