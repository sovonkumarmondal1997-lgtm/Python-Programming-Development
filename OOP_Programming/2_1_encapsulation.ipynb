```python
"""
Write OOP classes to handle the following scenarios:
A user can create and view 2D coordinates
A user can find out the distance between 2 coordinates
A user can find find the distance of a coordinate from origin
A user can check if a point lies on a given line
A user can find the distance between a given 2D point and a given line
"""
```

```python
import math 

class Point:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def view_point(self):
        print(f'({self.x}, {self.y})')

    def distance_between_points(self, other_point):
        distance = math.sqrt((self.x - other_point.x)**2 + (self.y - other_point.y)**2)
        return distance
    
    def distance_from_origin(self):
        distance = math.sqrt(self.x**2 + self.y**2)
        return distance
    
class Line:

    def __init__(self, A, B, C):
        self.A = A
        self.B = B
        self.C = C

    def view_line(self):
        print(f'Line: {self.A}x + {self.B}y + {self.C} = 0')

    def check_point_on_line(self, point):
        result = self.A * point.x + self.B * point.y + self.C
        if result == 0:
            return True
        else:
            return False
        
    def distance_from_point(self,point):
        numerator = abs(self.A * point.x + self.B * point.y + self.C)
        denominator = math.sqrt(self.A**2 + self.B**2)
        return numerator / denominator
```

```python
p1 = Point(3, 4)
p2 = Point(6, 8)

p1.view_point()
p2.view_point()

print("Distance between p1 and p2:", p1.distance_between_points(p2))
print("Distance of p1 from origin:", p1.distance_from_origin())


line1 = Line(2, 3, -12)
line1.view_line()

print("Does p1 lie on line?", line1.check_point_on_line(p1))
print("Distance between p1 and line:", line1.distance_from_point(p1))
```

```python
class Person:

  def __init__(self,name_input,country_input):
    self.name = name_input
    self.country = country_input

  def greet(self):
    if self.country == 'india':
      print('Namaste',self.name)
    else:
      print('Hello',self.name)
```

```python
person1 = Person('John','india')
person1.greet()
```

```python
# How to access attributes:
p = Person('Sovon', 'india')
p.name
```

```python
# How to access methods:
p.greet()
```
