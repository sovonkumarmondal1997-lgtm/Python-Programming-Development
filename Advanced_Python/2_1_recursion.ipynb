```python
# Recursion in Python means a function calls itself again and again until a stopping condition is reached.

def countdown(n):
    if n == 0:
        print("Done")
    else:
        print(n)
        countdown(n - 1)
```

```python
countdown(5)

# Here, countdown() keeps calling itself, and n == 0 is the stopping condition, called the base case.
```

```python
# Example 1: Factorial:
# Factorial means: 5! = 5 × 4 × 3 × 2 × 1

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
print(factorial(5))

'''
factorial(5)
= 5 * factorial(4)
= 5 * 4 * factorial(3)
= 5 * 4 * 3 * factorial(2)
= 5 * 4 * 3 * 2 * factorial(1)
= 5 * 4 * 3 * 2 * 1
= 120

Here, n == 1 is the base case.
'''
```

```python
# Example 2: Sum of numbers:
# sum(5) = 5 + 4 + 3 + 2 + 1

def total_sum(n):
    if n == 1:
        return 1
    return n + total_sum(n - 1)
```

```python
print(total_sum(5))

'''
total_sum(5)
= 5 + total_sum(4)
= 5 + 4 + total_sum(3)
= 5 + 4 + 3 + total_sum(2)
= 5 + 4 + 3 + 2 + total_sum(1)
= 5 + 4 + 3 + 2 + 1
= 15

Here also, n == 1 is the base case.
'''
```

```python
'''
Main idea: recursion needs two things:
1. Base case → where the function stops
2. Recursive case → where the function calls itself
'''
```
