```python
"""
Serialization and Deserialization:

1. Serialization - process of converting python data types to JSON format
2. Deserialization - process of converting JSON to python data types

3. what is JSON:
JSON is a lightweight text format used to store and transfer data in key-value pairs. JSON full form is JavaScript Object Notation. JSON is a language independent data format and it is human readable. JSON is used to store and transfer data in web applications. JSON is also used to store data in files and databases.
"""
```

```python
# serialization using json module: list to json
import json

L = [1,2,3,4,5]

with open('sample.json', 'w') as f:
    json.dump(L, f, indent = 4) # indent is used to make the json file more readable
```

```python
# serialization using json module: dictionary to json
dic = {'name':'Selmon Bhoi','age':60,'city':'Mumbai'}
with open('sample1.json', 'w') as f:
    json.dump(dic, f, indent = 4) # indent is used to make the json file more readable
```

```python
# deserialization using json module: json to dictionary
with open('sample1.json', 'r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# serialization and deserialization of tuples using json module:
tup = (1,2,3,4,5)
with open('sample3.json', 'w') as f:
    json.dump(tup, f, indent = 4)

with open('sample3.json', 'r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# serialization and deserialization of a nested dictionary with json module:

nested_dic = {'name':'Selmon Bhoi','age':60,'city':'Mumbai','hobbies':['acting','dancing','singing'],'education':{'graduation':'B.A.','post_graduation':'M.A.'}}
with open('sample4.json', 'w') as f:
    json.dump(nested_dic, f, indent = 4)

with open('sample4.json', 'r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# serializing and deserializing custom objects using json module:

class Person:

  def __init__(self,fname,lname,age,gender):
    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender

# format to printed in
# -> Nitish Singh age -> 33 gender -> male
```

```python
person = Person('Nitish','Singh',33,'male')
```

```python
# As a string
import json

def show_object(person):
  if isinstance(person,Person):
    return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

with open('demo.json','w') as f:
  json.dump(person,f,default=show_object)
```

```python
# As a dict
import json

def show_object(person):
  if isinstance(person,Person):
    return {'name':person.fname + ' ' + person.lname,'age':person.age,'gender':person.gender}

with open('demo.json','w') as f:
  json.dump(person,f,default=show_object,indent=4)
```

```python
# deserializing
import json

with open('demo.json','r') as f:
  d = json.load(f)
  print(d)
  print(type(d))
```

```python
"""
1. Pickling: Pickling is the process of converting a Python object into a binary file format so it can be saved and loaded later.

2. Unpickling: Unpickling is the process of loading binary pickled data back into the original Python object.
"""
```

```python
class Person:

  def __init__(self,name,age):
    self.name = name
    self.age = age

  def display_info(self):
    print('Hi my name is',self.name,'and I am ',self.age,'years old')
```

```python
p = Person('nitish',33)
```

```python
# pickle dump
import pickle
with open('person.pkl','wb') as f:
  pickle.dump(p,f)
```

```python
# pickle load
import pickle
with open('person.pkl','rb') as f:
  p = pickle.load(f)

p.display_info()
```

```python
"""
Pickling vs JSON:

1. Pickling is used in Python to save Python objects like lists, dictionaries, classes, and custom objects into a binary format. It is mainly used when you want to store Python-specific data and load it back later in the same Python environment.

2. JSON is used to store and transfer data in a text format using key-value pairs. It is language-independent, so data saved as JSON can be used by Python, JavaScript, Java, APIs, websites, and many other systems.

3.Main difference: Pickling is Python-specific and stores data in binary format, while JSON is language-independent and stores data in readable text format.
"""
```
