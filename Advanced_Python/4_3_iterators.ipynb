```python
"""
1. iteration:
    Iteration in Python means repeatedly going through each item of a sequence or iterable, such as a list, tuple, string, or range, one by one.
    
"""
```

```python
# End-to-end example of iteration in Python

students = [
    {"name": "Amit", "marks": 85},
    {"name": "Riya", "marks": 72},
    {"name": "Sovon", "marks": 90},
    {"name": "Neha", "marks": 60}
]

total_marks = 0

for student in students:
    print(f"Student: {student['name']}, Marks: {student['marks']}")
    total_marks += student["marks"]

average = total_marks / len(students)

print("Total Marks:", total_marks)
print("Average Marks:", average)
```

```python
"""
2. iterators:
An iterator in Python is an object that remembers its position and returns the next item one at a time using next().
"""
```

```python
# End-to-end example of iterator in Python:

numbers = [10, 20, 30]

# Step 1: Convert iterable list into an iterator
num_iterator = iter(numbers)

# Step 2: Get values one by one using next()
print(next(num_iterator))  # 10
print(next(num_iterator))  # 20
print(next(num_iterator))  # 30

# Step 3: No item left, so this will raise StopIteration
print(next(num_iterator))
```

```python
# Iterable:
# An iterable in Python is any object that can be looped over using a for loop, such as a list, tuple, string, dictionary, or set.

# End-to-end example of iterable in Python

fruits = ["Apple", "Mango", "Banana"]

# 'fruits' is an iterable object
for fruit in fruits:
    print(fruit)
```
