```python
"""
Exceptions:

1. Exceptions happen while the program is running.
2. They happen when something unexpected occurs.
3. Exceptions are raised by the Python runtime.
4. You have to handle them while the program is running.

Examples:

1. Memory overflow
2. Dividing by zero
3. Database error

"""
```

```python
"""
Handling exceptions in Python is important because:

1. It prevents program crashes
If an error happens, the program can continue instead of stopping suddenly.

2. It gives better error messages
You can show a clear message instead of a confusing Python error.

3. It protects important work
Example: closing files, saving data, or disconnecting from a database properly.

4. It handles unexpected situations
Example: wrong user input, missing file, internet problem, database error.

5. It makes the program reliable
A good program should not break easily when something goes wrong.

"""
```

```python
# We can handle exceptions in Python using try and except blocks.
```

```python
# create a new file and write some text in it:
with open('test.txt', 'w') as f:
    f.write('Hello, world! This is a test file.')
```

```python
# without exception handling if we try to open the file in read mode, it will work fine, but if there is any 
# problem with the file, it will raise an error and crash the program. For example, if we try to open a non-existent file:

with open ('test11.txt', 'r') as f:
    content = f.read()
    print(content)

# This will raise a FileNotFoundError and crash the program. But if we handle the exception, we can show a better message and prevent the crash:
```

```python
# lets try to open the file again but this time with exception handling:

try:
    with open ('test11.txt', 'r') as f:
        content = f.read()
        print(content)
except:
    print('Sorry, file not found. Please check the file name and try again.')
```

```python
# another example of exception handling:
try:
    with open('test.txt', 'r') as f:
        content = f.read()
        print(content)
        print(m)
except:
    print('Sorry, some error occurred.')

# this is the example of sarkari coding
```

```python
# catching specific exception error: to check what type of error happening in your code:
try:
    with open('test.txt', 'r') as f:
        content = f.read()
        print(content)
        print(m)
except Exception as e:
    print('Error details:', e)
```

```python
# catching specific exception error: to check what type of error happening in your code with name and traceback details:
try:
    with open('test.txt', 'r') as f:
        content = f.read()
        print(content)
        print(m)
except Exception as e:
    print('Error details:', e.with_traceback)
```

```python
# final example of exception handling with traceback details:
try:
    m = 10
    with open('test.txt', 'r') as f:
        content = f.read()
        print(content)
        print(m)
        print(10/2)
        l = [1, 2, 3]
        l = l[5]
except NameError:
    print('Sorry, a NameError occurred. Please check your code and try again.')
except FileNotFoundError:
    print('Sorry, file not found. Please check the file name and try again.')
except ZeroDivisionError:
    print('Sorry, a ZeroDivisionError occurred. Please check your code and try again.')
except Exception as e: # this is a generic exception handler that will catch any error that is not caught by the previous except blocks
    print('Error details:', e)
    
```

```python
# else: In a try-except block, we use else to run code only when no exception happens.

try:
    f = open('test.txt', 'r')
except FileNotFoundError:
    print('Sorry, file not found. Please check the file name and try again.')
except Exception as e:
    print('Error details:', e)
else:
    print(f.read())
    print(f.close())
```

```python
# finally : We use finally to run code always. Usually, finally is used for cleanup work, like closing files or database connections.

try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ZeroDivisionError:
    print("Cannot divide by zero")

else:
    print("Result is:", result)

finally:
    print("Program finished")
```

```python
# finally:
try:
    file = open("data.txt", "r")
    content = file.read()

except FileNotFoundError:
    print("File not found.")

else:
    print(content)

finally:
    print("Closing file if it was opened.")
```
