```python
"""
Theory:
Types of data used for I/O (Input/Output) operations with users:

1. Text - '12345' as a sequence of unicode characters
2. Binary - 12345 as a sequence of bytes of its binary equivalent

Hence there are 2 file types to deal with:

1. Text files - All program files are text files
2. Binary Files - Images,music,video,exe files
"""
```

```python
"""
How File I/O is done in most programming languages:

1. Open a file
2. Read/Write data
3. Close the file
"""
```

```python
# Writing to a file:

# case 1: file is not created and present in the current directory:
f = open('samplefile.txt', 'w') # it will create a new file named samplefile.txt in the current directory and open it in write mode ('w')
f.write('This is file 1') # it will write the string 'This is file 1' to the file samplefile.txt
f.close() # it will close the file samplefile.txt

# case 2: write multiple lines to a file:
a = open('samplefile_1.txt', 'w')
a.write("This is the first line of the file.\n") # \n is used to create a new line
a.write("This is the second line of the file.\n")
a.write("This is the third line of the file.")
a.close()
```

```python
# case 3: write strings to a existing file:
b = open('samplefile.txt', 'a') # it will open the existing file samplefile.txt in append mode ('a')
b.write('\nThis is file 2') # it will write the string 'This is file 2' to the file samplefile.txt in a new line
b.close() # it will close the file samplefile.txt

# if we want to write stings to a existing file we can use ('w') mode but it will overwrite the existing content of the file and write the new content to the file. Hence we should use ('a') mode to write stings to a existing file without overwriting the existing content of the file.
```

```python
# case 4: write a list of strings to a file:
lines = ['This is the first line of the file.\n', 'This is the second line of the file.\n', 'This is the third line of the file.']
c = open('samplefile_2.txt', 'w')
c.writelines(lines) # it will write the list of strings to the file samplefile_2.txt
c.close()
```

```python
# Read from a file:
# case 1: read the content of a file:
f = open('samplefile.txt', 'r') # it will open the file samplefile.txt in read mode ('r')
s = f.read() # it will read the content of the file samplefile.txt and store it in the variable s
print(s) # it will print the content of the file samplefile.txt
f.close()
```

```python
# case 2: read a file till n characters:
f = open('samplefile.txt', 'r')
s = f.read(20) # it will read the first 20 characters of the file samplefile.txt and store it in the variable s
print(s) # it will print the first 20 characters of the file samplefile.txt
f.close()
```

```python
# case 3: read a file line by line:
f = open('samplefile_1.txt', 'r')
s = f.readline() # it will read the first line of the file samplefile_1.txt and store it in the variable s
t = f.readline() # it will read the second line of the file samplefile_1.txt and store it in the variable s
print(s) # it will print the first line of the file samplefile_1.txt
print(t) # it will print the second line of the file samplefile_1.txt
f.close()
```

```python
# case 4: reading entire file line by line:
f = open('samplefile_1.txt', 'r')
for line in f: # it will read the file samplefile_1.txt line by line and store each line in the variable line
    print(line) # it will print each line of the file samplefile_1.txt
f.close()
```

```python
"""
Using Context Manager (With): 

1. It's a good idea to close a file after usage as it will free up the resources
2. If we dont close it, garbage collector would close it
3. with keyword closes the file as soon as the usage is over
4. It is a good practice to use with keyword while working with files as it will ensure that the file is closed after usage and it will also handle any exceptions that may occur while working with files.
5. we will use with keyword insted of close() method to close the file after usage.
"""
```

```python
# working with files using with keyword: write the file
with open('samplefile.txt', 'w') as f:
    f.write('Selmon Bhoi') # it will write the string 'Selmon Bhoi' to the file samplefile.txt
# it will automatically close the file samplefile.txt after the usage of the with block is over
```

```python
# working with files using with keyword: read the file
with open('samplefile.txt', 'r') as f:
    s = f.read() # it will read the content of the file samplefile.txt and store it in the variable s
    print(s) # it will print the content of the file samplefile.txt
```

```python
# moving within a file: 10 character then 10 character
with open('samplefile_1.txt', 'r') as f:
    s = f.read(10) # it will read the first 10 characters of the file samplefile_1.txt and store it in the variable s
    print(s) # it will print the first 10 characters of the file samplefile_1.txt
    t = f.read(10) # it will read the next 10 characters of the file samplefile_1.txt and store it in the variable t
    print(t) # it will print the next 10 characters of the file samplefile_1.txt

# this is happening for the buffer memory.
# using this concept we can load a huge file in chunks and process it without loading the entire file into the memory at once.
```

```python
# creating a big filein the memory:
big_L = ['hello world ' for i in range(1000)]

with open('big.txt','w') as f:
  f.writelines(big_L)
```

```python
# load the big.txt file in chunks and process it without loading the entire file into the memory at once:
with open('big.txt','r') as f:

  chunk_size = 10

  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size),end='***')
    f.read(chunk_size)
```

```python
# 1. seek(): seek() moves the file cursor to a specific position in a file so Python can read or write from that location.

# 2. tell(): tell() returns the current position of the file cursor in a file.

with open('samplefile_1.txt','r') as f:
    print(f.read(10))
    print(f.tell())
    print(f.read(10))
    print(f.tell())

    print(f.seek(0))
    print(f.read(10))
    print(f.tell())
```

```python
# working with seek() during write:
with open('samplefile.txt','w') as f:
    f.write('Hello Selmon Bhoi')
    f.seek(0)
    f.write('Ha')
    
```
