```python
"""
Raise Exception:
Raise exception in Python means manually creating an error using the raise keyword when something invalid happens. This is useful when you want to enforce certain conditions in your code and provide a clear error message when those conditions are not met.
"""
```

```python
age = 15

if age < 18:
    raise ValueError("Age must be 18 or above")

"""
1. raise manually creates an exception.
2. Here, if age is less than 18, Python raises a ValueError.
3. The program stops and shows the error message:
"""
```

```python
class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise Exception('amount cannot be -ve')
    if self.balance < amount:
      raise Exception('paise nai hai tere paas')
    self.balance = self.balance - amount

obj = Bank(10000)
try:
  obj.withdraw(15000)
except Exception as e:
  print(e)
else:
  print(obj.balance)
```

```python
# creating custom exception class:
# We create our own custom exception class in Python to show a specific error for our own program/business rule.

class MyException(Exception):
  
  def __init__(self,message):
    print(message)

class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise MyException('amount cannot be -ve')
    if self.balance < amount:
      raise MyException('paise nai hai tere paas')
    self.balance = self.balance - amount

obj = Bank(10000)
try:
  obj.withdraw(5000)
except MyException as e:
  pass
else:
  print(obj.balance)
```
