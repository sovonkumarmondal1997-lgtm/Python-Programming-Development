```python
"""
1. Generators:
   A generator in Python is a special iterator that produces values one at a time using yield, saving memory by not storing all values at once.

2. Why we use gnerators:
   We use generators in Python to produce values one at a time, which saves memory and improves performance when working with large data. 
"""
```

```python
# End-to-end example of generator in Python

def count_up_to(n):
    count = 1

    while count <= n:
        yield count
        count += 1


numbers = count_up_to(5)

print(next(numbers))  # 1
print(next(numbers))  # 2

for num in numbers:
    print(num)
```
