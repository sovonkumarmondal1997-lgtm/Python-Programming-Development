```python
"""
A program can have errors in 2 stages:

1. Before running the program → Syntax Error
   Example: wrong spelling, missing colon, wrong brackets.
2. While running the program → Exception
   Example: dividing by zero, file not found, wrong input type.

"""
```

```python
"""
Syntax Error means the program is written incorrectly.

The interpreter/compiler finds this error before the program runs.

You fix it by correcting the code.
"""
```

```python
# Examples of syntax error
print 'hello world'
```

```python
"""
Other examples of Syntax Error:

1. Missing colon or brackets
2. Spelling a keyword wrongly
3. Wrong indentation
4. Writing empty if, else, loop, class, or function blocks
"""
```

```python
# example of syntax error: missing colon
a = 5
if a==3
  print('hello')
```

```python
# example of syntax error: wrong spelling
a = 5
iff a==3:
  print('hello')
```

```python
# example of syntax error: wrong indentation (special syntax error in Python)
a = 5
if a==3:
print('hello')
```
