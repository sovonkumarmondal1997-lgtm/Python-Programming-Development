```python
"""
Decorators:
-> A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it.

This can happen only because python functions are 1st class citizens.

There are 2 types of decorators available in python:

1. Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
2. User defined decorators that we programmers can create according to our needs

"""
```

```python
# the foundation of decorators is A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it.

def modify(func,number):
    return func(number)

def squre(num):
    return num**2
```

```python
modify(squre,5)
```

```python
# simple example of decorators: this example decribed the functionality of decorators in a simple way

def my_decorator(func):

    def wrapper():
        print("**********")
        func()
        print("**********")
    return wrapper

def hello():
    print("Hello World")

def display():
    print("Welcome to Python Programming")
```

```python
a = my_decorator(hello)
a()
```

```python
b = my_decorator(display)
b()
```

```python
# Better syntax for using decorators is to use @ symbol before the function definition that we want to decorate:

def my_decorator(func):

    def wrapper():
        print("**********")
        func()
        print("**********")
    return wrapper

@ my_decorator
def hello():
    print("Hello World")
@ my_decorator
def display():
    print("Welcome to Python Programming")
```

```python
hello()
```

```python
display()
```

```python
# example of decorators:
import time

def timer(func):

    def wrapper():
        start = time.time()
        func()
        print(f"Time taken to execute the function is {time.time() - start} seconds")
    return wrapper

@timer
def hello():
    print("Hello World")
    time.sleep(2)

# in this program only pronblem is that we are not able to pass any argument to the function that we want to decorate, so we will use *args and **kwargs to solve this problem
```

```python
hello()
```

```python
# example of decorators: with *args
import time

def timer(func):

    def wrapper(*args):
        start = time.time()
        func(*args)
        print(f"Time taken to execute the function is {time.time() - start} seconds")
    return wrapper

@timer
def hello():
    print("Hello World")
    time.sleep(2)

@timer
def squre(num):
    print(num**2)
    time.sleep(2)
```

```python
hello()
squre(5)
```

```python
def sanity_check(data_type):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(*args) == data_type:
        func(*args)
      else:
        raise TypeError('This datatype is not allowed')
    return inner_wrapper
  return outer_wrapper

@sanity_check(int)
def square(num):
  print(num**2)

@sanity_check(str)
def greet(name):
  print('hello',name)
```

```python
square(10)
```

```python
greet("Python")
```

```python
# example of a decorator that takes arguments: *kwargs
def my_decorator(func):
    def wrapper(**kwargs):
        print("Before function call")
        func(**kwargs)
        print("After function call")
    return wrapper


@my_decorator
def greet(name, age):
    print(f"My name is {name} and I am {age} years old")
```

```python
greet(name="Sovon", age=25)
```
