```python
"""
1. Namespaces:
-> A namespace is a space that holds names(identifiers).Programmatically speaking, namespaces are dictionary of identifiers(keys) and their objects(values)

a = 2, b = 3, c = 4
Here a, b, c are identifiers and 2, 3, 4 are their corresponding objects. So we can say that a, b, c are keys and 2, 3, 4 are values in the namespace dictionary.

2. There are 4 types of namespaces:

1. Builtin Namespace
    Contains Python's predefined names like print(), len(), sum(), and type().
2. Global Namespace
    Contains names created at the main program/file level, outside all functions.
3. Enclosing Namespace
    Contains names from an outer function when functions are nested.
4. Local Namespace
    Contains names created inside the current function.

3. Scope:
Scope means the area of your Python code where a variable/name can be used directly.

4. LEGB Rule:
Python searches for a variable using the LEGB rule:

L — Local: inside the current function
E — Enclosing: inside the outer function, if functions are nested
G — Global: outside all functions, in the main file
B — Built-in: Python's built-in names like print, len, sum, etc.
"""
```

```python
# 1. LEGB Rule: local and global

a = 2 # a is a global variable which is created in the global namespace and present in global scope

def temp():
  b = 3 # b is a local variable which is created in the local namespace of temp() function and present in local scope of temp() function
  print(b)
```

```python
temp()
print(a)
```

```python
# 2. local and global -> same name
# both global and local variable can have same name. In that case, local variable will shadow the global variable inside the function and we will get the value of local variable when we access that variable inside the function. both local and global variable will be present in their respective namespaces and scopes.

a = 2

def temp():
  a = 3
  print(a)
```

```python
temp()
print(a)
```

```python
# 3. local and global -> local does not have but global has
# local can access global variable but global cannot access local variable as per LEGB rule. If local variable is not present in the local namespace but global variable is present in the global namespace, then we will get the value of global variable when we access that variable inside the function.

a = 2

def temp():
  print(a)
```

```python
temp()
print(a)
```

```python
# 4. local and global -> editing global
# we can edit global variable inside the function but we need to use global keyword to tell Python that we want to edit the global variable but it is not a good programming practice we should avoid it. If we do not use global keyword and try to edit the global variable, then Python will create a new local variable with the same name and we will get UnboundLocalError because we are trying to access the local variable before assigning any value to it.

a = 2

def temp():
  global a
  a += 1
  print(a)
```

```python
temp()
print(a)
```

```python
# 5. local and global -> a global variable can be created inside local variable

def temp():
  global a
  a = 1
  print(a)
```

```python
temp()
print(a)
```

```python
# 6. local and global -> function parameter is local
# In this code, z is a local variable because it is a function parameter and exists only inside the temp() function.

def temp(z):
  print(z)
```

```python
a = 5
temp(5)
print(a)
```

```python
# Built - in scope:
# Built-in scope is the outermost scope that contains Python’s predefined names like print(), len(), type(), and sum(), which are available everywhere in your program.

# to check wgat are the built in functions are available in built-in scope, we can use dir() function on builtins module which contains all the built-in names.

import builtins
print(dir(builtins))
```

```python
# renaming built-ins:
# This program first calls Python’s built-in max() to print 3, then defines a custom function named max, which hides the built-in max() and causes a TypeError when max(L) is called again.

L = [1,2,3]
print(max(L))

def max():
  print('hello')

print(max(L))
```

```python
# Enclosing scope:

def outer():
  def inner():
    print(a)
  inner()
  print('outer function')
```

```python
a = 1
outer()
print('main program')
```

```python
# nonlocal keyword: we can change outer variable from inner function using nonlocal keyword. If we do not use nonlocal keyword and try to edit the outer variable, then Python will create a new local variable with the same name and we will get UnboundLocalError because we are trying to access the local variable before assigning any value to it.

def outer():
  a = 1
  def inner():
    nonlocal a
    a += 1
    print('inner',a)
  inner()
  print('outer',a)
```

```python
outer()
print('main program')
```
