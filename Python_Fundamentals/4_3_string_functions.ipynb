```python
"""
Common functions: Works on all data types in python 
1. length (len)
2. maximum (max)
3. minimum (min)
4. sorted (sorted)
5. sum (sum)
"""
```

```python
# len: count the length of the string including 'space'
len('hello world')
```

```python
# max: extract the maximum_value character from the string based on it's ascii value
max('hello kolkata')
```

```python
# min: extract the minimum_value character from the string based on the ascii value
min('hello kolkata')

# here 'space' ascii value is the lowest ascii value in the entire string
```

```python
# sorted: all characters sorted as per ascii value from minimum to maximum (ascending order)
sorted('hello kolkata')
```

```python
# sorted: all characters sorted as per ascii value from maximum to minimum (descending order)
sorted('hello kolkata',reverse=True)
```

```python
# sum: add all the total ascii values of the string
# sum do not work directly on string

# Summing ASCII (Unicode) Values of Characters
text = "hello"
result = sum(ord(char) for char in text)
print(result)

# ord(char) converts each character into its corresponding ASCII value.
```

```python
# Summing Digits in a Numeric String

numbers = "12345"
result = sum(int(digit) for digit in numbers)
print(result)
```

```python
# Using sum() with Boolean Conditions on Strings: count special characters

text = "dataengineering"
vowel_count = sum(char in "aeiou" for char in text.lower())
print(vowel_count)
```
