```python
# TIME COMPLEXITY OF n^2 + 2n + 2
# =================================

# When we have a formula like n^2 + 2n + 2, we use Big O to describe how it grows
# as n becomes large. Big O keeps only the fastest-growing term and ignores
# constant factors.

# Step 1: Identify the terms:
#   n^2    -> quadratic term (dominant term)
#   2n     -> linear term (multiplicative constant 2 doesn't affect growth rate)
#   2      -> constant term (additive constant doesn't affect growth rate)

# Step 2: Keep the dominant term as n grows large.
# The quadratic term n^2 grows much faster than 2n or 2, so it dominates.

# Step 3: Drop lower-order terms and constants.
#   n^2 + 2n + 2  =>  O(n^2)

# So the time complexity is O(n^2).

# HOW TO MEASURE IT BY COUNTING OPERATIONS
# ========================================
# If this expression comes from code, it usually reflects an algorithm with:
# - one nested loop over n items (n^2 work)
# - some extra single loops or operations (2n and constant work)

# Example: a double loop plus a simple extra pass:
# for i in range(n):            # n iterations
#     for j in range(n):        # another n iterations inside -> n * n = n^2
#         do_constant_work()
# for k in range(n):            # another n iterations -> 2n when combined
#     do_constant_work()
# do_constant_work()             # constant work -> +2 is similar to a few constant steps

# As n grows, the n^2 term is the one that controls runtime.
# That is why the time complexity is O(n^2), even though the exact count is n^2 + 2n + 2.

# PRACTICAL MEASUREMENT TIP:
# Use theoretical Big O for algorithm comparison.
# Use timing tools like timeit only to compare actual runtime on your machine.
```
