```python
"""
To master the difference between Typing and Binding, you must remember that Typing is about the container (the variable),
while Binding was about the action (the method).
"""
```

```python
"""
Dynamic Binding (Late Binding)
Dynamic binding is when the computer waits until the program is actually running to decide which method to execute.
It looks at the actual object created in memory, not the variable type.

"""

# 1. We define two different types of objects
class CreditCard:
    def pay(self):
        print("Processing credit card payment...")

class Cash:
    def pay(self):
        print("Taking cash from wallet...")

# 2. We create a variable 'my_payment'
# We can change what 'my_payment' is while the program runs
my_payment = CreditCard()
my_payment.pay()  # The computer "binds" this to CreditCard's logic

my_payment = Cash()
my_payment.pay()  # The computer "binds" this to Cash's logic at runtime
```

```python
"""
Static Binding:
Static Binding is when the computer knows exactly which instruction to follow the moment you finish writing the code
(at Compile-time).The decision is "static" because it is fixed and cannot change while the program is running.
"""
# 1. We define a class with a static method in Java language 
class Calculator {
    // This is a static method
    static void compute() {
        System.out.println("Performing math...");
    }
}

public class Main {
    public static void main(String[] args) {
        // The computer knows EXACTLY which 'compute' to run 
        // just by looking at the word 'Calculator'.
        Calculator.compute(); 
    }
}
```

```python
"""
1. Static Typing (The "Fixed Label" Approach)
In Static Typing, the data type of a variable is determined at compile-time.
Once you label a variable as an "Integer," it can never hold anything else for the rest of its life
"""
JAVA Code:

int x = 10;      // x is now a permanent 'int' jar
x = "Hello";     // ERROR! The compiler stops you before the program even runs.
```

```python
"""
Dynamic Typing (The "Sticky Note" Approach)
Definition: In Dynamic Typing, the data type is checked at runtime.
Variables don't have a fixed type; only the values have types.
You can change what is inside the variable at any time
"""
x = 10           # x currently points to an integer
print(type(x))   # <class 'int'>

x = "Hello"      # No problem! x now points to a string
print(type(x))   # <class 'str'>
```
