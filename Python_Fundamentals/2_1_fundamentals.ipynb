```python
print('Hello',1,7.898,True)

# between every data type there is a 'space'. it is happening for a default print setting,
# which we can see in the manual of print module. that is sep=''
```

```python
print('Hello',1,7.898,True,sep='/')

# here we change the default setting from sep=' ' to sep='/' only for this reason when we print, 
# all the datatypes together we can see '/' between them 

# using sep=' ' we can use anything to print the desired result
```

```python
print("Hello")
print("World")

# we can see different print module show the result as output in different line not in a single line

# in print module there is a default option is end='\n', \n = line change, that is the reason,
# when we use print module in different line it shows output in different line not in a single line
```

```python
print("Hello",end='-')
print("World")

# here we overwrite the default value '\n' with '-' thats why it is not changing the line,
# and showing output in one line with '-'
```
