```python
# Find the length of a given string without using the len() function:

# Take the input from the user
string_input = input("Enter the string: ")

# Take the length value 
counter = 0

# Length logic:
for i in string_input:
    counter = counter + 1

# Display the result:
print("The length of the string is:",counter)
```

```python
# Extract the username from a given mail.
# example-> Email: sovon9988@gmail.com, Username: sovon9988

# Take the email as input from the user
email = input('Enter the email:')

# Logic of the program
position = email.index('@')
username = email[0:position]

# Display the result
print("Username:",username)
```

```python
# count the frequency of a particular character in a provide string
# Ex-> string: "hi how are you?", frequency of h: 2, frequency of a: 1

# Take the input from the user
str = input("Enter the string: ")
term = input("Enter the term to find the frequency: ")

# Set a variable to store the counter
counter = 0

# Logic of the program
for i in str:
    if i == term:
        counter = counter + 1

# Display the result
print(counter)
```

```python
# write a program which can remove a particular character from a string:

# Take an input from the user
str = input("Enter the string: ")
term = input("Enter the character would you like to remove: ")

# set a variable to store the string after iteration
result = ""

# Logic of the program
for i in str:
    if i != term:
        result = result + i

# Display the result
print(result)
```

```python
# write a program which can remove a particular character from a string:

# Take an input from the user
str = input("Enter the string: ")
term = input("Enter the character would you like to remove: ")

# set a variable to store the string after iteration
result = ""

# Logic of the program
for i in str:
    if i == term:
        continue
    result = result + i

# Display the result
print(result)
```

```python
# write a program that can check wheather a given string is palindrome or not:

str = input('Enter the String: ')

rev = str[::-1]

if str == rev:
    print("It is a palindrome, string value:{0}, result:{1}".format(str,rev))
else:
    print("It is not a palindrome, string value:{0}, result:{1}".format(str,rev))
```

```python
# write a program to count the number of words in a string without split():

str = input("Enter the string: ")
temp = ''
lst = []

for i in str:
    if i != ' ':
        temp = temp + i
    else:
        lst.append(temp)
        temp = ''

lst.append(temp)
print(lst)
```

```python
# Write a python program to convert a string to title case without using the title():
str = input("Enter the string: ")

L = []

for i in str.split():
    L.append(i[0].upper() + i[1:].lower())

print(" ".join(L))
```

```python
# write a program that can convert an integer into a string without str() function:

number = int(input("Enter the number: "))

digits = '0123456789'
result = ''

while number != 0:
    result = result + digits[number % 10]
    number = number // 10

print(result)
print(type(result))
```
