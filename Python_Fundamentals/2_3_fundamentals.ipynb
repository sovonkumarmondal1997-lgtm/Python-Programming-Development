Type Conversion:

```python
# 1. Implicit type conversion: Python automatically converts data types during operations.

c = 5
d = 7.12
print(c + d)

# 2. Explicit type conversion: You manually change data type using functions.

a = int("5")
v = 5
print(a + v)
print(type(a))
```

```python
# python type conversion does not change the original data, its create a new value 
# and we perform the task upon this new value 
```

```python
# Literals:
# -> whatever raw value we save in variables thats called literals 

# Binary Literals:
a = 0b1010
print(a)

# Decimal Literals:
b = 100
print(b)

# Octal Literals:
c = 0o310
print(c)

# Hexadecimal Literals:
d = 0x12c
print(d)
```

```python
# Float Literals:

flo_1 = 10.5
print(flo_1)

flo_2 = 1.5e2
print(flo_2)

flo_3 = 1.5e-3
print(flo_3)
```

```python
# Complex Literals:

x = 3.14j
print(x)
print(x.real, x.imag)
```

```python
# Different types of represent strings:

string = 'this is a string'
print(string)

strings = "this is a stings"
print(strings)

single_character = "C"
print(single_character)

multiline_string = """This is a multiline string"""
print(multiline_string)

unicode = u"\U0001f600\U0001f606\U0001f923"
print(unicode)

raw_string = r"raw \n string"
print(raw_string)
```
