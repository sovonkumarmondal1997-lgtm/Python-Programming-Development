```python
# Accessing substrings from strings: Indexing
```

```python
# Positive indexing: access single substring
s = "hello world"
print(s[0])
print(s[1])
print(s[2])
print(s[3])
print(s[4])
print(s[5])
print(s[6])
print(s[7])
print(s[8])
print(s[9])
print(s[10])
```

```python
# Negative indexing: access single substring
s = "hello world"
print(s[-1])
print(s[-2])
print(s[-3])
print(s[-4])
print(s[-5])
print(s[-6])
print(s[-7])
print(s[-8])
print(s[-9])
print(s[-10])
print(s[-11])
```

```python
# Positive slicing: access multiple substring
s = "hello world"
print(s[0:5])
print(s[2:5])
print(s[4:5])
print(s[2:])
print(s[:5])

# accessing full string at once:
print(s[:])

# accessing string through step size:
print(s[0:11:2])
print(s[0:11:3])
```

```python
# slicing with negative value:
s = "hello world"
print(s[11:0:-2])

# fully reverse a string with negative value
print(s[::-1])
```

```python
# Negative slicing: access multiple substring
s = "hello world"
print(s[-5:])
print(s[-1:-6:-1])
print(s[-11:])
```
