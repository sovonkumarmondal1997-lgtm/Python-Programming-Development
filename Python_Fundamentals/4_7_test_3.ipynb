```python
"""
Problem 5: Write a Python Program to Find the Sum of the Series till the n_th term:
1 + x^2/2 + x^3/3 + … x^n/n
n will be provided by the user
x will be provided by the user
"""
x = int(input("Enter the value of (x): "))
n = int(input("Enter the value of (n): "))
total_sum = 1
equation = ''

for i in range(2,n+1):
    total_sum = total_sum + (x ** i) / i
    equation = equation + "x^{}/{}+".format(i,i)

print(equation[:-1])
print(round(total_sum,2))
```

```python
# ln(x)= (x-1)/x + 1/2 * ((x-1/x))**2 + 1/3 * ((x-1/x))**3 + 1/4 * ((x-1/x))**4 + 1/x * ((x-1/x))**n

x = int(input("Enter the value of (x): "))
n = int(input("Enter the value of (n): "))

total_sum = 0
terms_list = []

# Pre-calculate the base once to save CPU cycles
# Instead of doing the math (x-1)/x every single loop iteration
base = (x - 1) / x 

for i in range(1, n + 1):
    # Calculation
    term_value = (1 / i) * (base ** i)
    total_sum += term_value
    
    # Building the visual representation correctly
    # We use (x-1)/x in the string to match the calculation
    terms_list.append("(1/{})*((x-1)/x)**{}".format(i, i))

# Join all parts with a + sign
full_equation = " + ".join(terms_list)

print(f"Approximation Equation: {full_equation}")
print("-" * 30)
print(f"Result: {round(total_sum, 2)}")
```

```python
# Problem 9: Write a program that will take a decimal number as input
# and prints out the binary equivalent of the number

n = int(input("Enter the number: "))
binary = []

while n > 0:
    binary.append(n % 2)
    n = n // 2
    print(n,binary)

for i in binary[::-1]:
    print(i,end='')
```

```python
# Problem 10: Write a program that will take 2 numbers as input and prints the LCM and HCF of those 2 numbers
# Standard Logic
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Keep copies of original numbers for LCM calculation later
x, y = a, b

# 1. Find HCF using the Euclidean Algorithm (most efficient)
while y != 0:
    x, y = y, x % y

hcf = x
# 2. Find LCM using the formula
lcm = (a * b) // hcf

print(f"HCF: {hcf}")
print(f"LCM: {lcm}")
```

```python
"""
Problem 11: Create Short Form from initial character
Given a string create short form of the string from Initial character. Short form should be capitalised.
Example:
Input: Data science mentorship program
Output: DSMP
"""
# Input the full string
phrase = input("Enter the string: ")

# 1. Split the string into a list of words
# 'Data science mentorship program' becomes ['Data', 'science', 'mentorship', 'program']
words = phrase.split()

short_form = ""

# 2. Iterate through each word
for word in words:
    # 3. Take the first character, capitalize it, and add to our string
    short_form = short_form + word[0].upper()

# 4. Print the final result
print("Output:", short_form)
```

```python
"""
Problem 12: Append second string in the middle of first string
Input:
campusx
data
Output: camdatapusx
"""

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

# 1. Calculate midpoint using floor division (//)
# If length is 7, mid will be 3
mid = len(s1) // 2

# 2. Slice the first string
# s1[:mid] gets everything BEFORE the middle
# s1[mid:] gets everything FROM the middle to the end
left_half = s1[:mid]
right_half = s1[mid:]

# 3. Combine them
result = left_half + s2 + right_half

print("Output:", result)
```

```python
"""
Problem 13:Given string contains a combination of the lower and upper case letters.
Write a program to arrange the characters of a string so that all lowercase letters should come first.
Given:
str1 = PyNaTive
Expected Output:
yaivePNT
"""

str1 = "PyNaTive"

lower = ""
upper = ""

# Iterate through each character
for char in str1:
    if char.islower():
        lower += char # Add to the lowercase bucket
    else:
        upper += char # Add to the uppercase bucket

# Combine them: lowercase first, then uppercase
result = lower + upper

print("Result:", result)
```

```python
"""
Problem 14:Take a alphanumeric string input and print the sum and average of the digits that appear in the string,
ignoring all other characters.
Input: hel123O4every093
Output:Sum: 22, Avg: 2.75
"""
user_input = "hel123O4every093"

total_sum = 0
count = 0

for char in user_input:
    # 1. Check if the character is a digit
    if char.isdigit():
        # 2. Convert to int and add to sum
        total_sum += int(char)
        # 3. Increment the counter
        count += 1

# 4. Final output logic
if count > 0:
    average = total_sum / count
    print(f"Sum: {total_sum}, Avg: {round(average, 2)}")
else:
    print("No digits found in the string.")
```

```python
"""
Problem 16: Check whether the string is Symmetrical.
Statement: Given a string. the task is to check if the string is symmetrical or not.
A string is said to be symmetrical if both the halves of the string are the same.

Example:
Input : khokho
Output : The entered string is symmetrical
"""
string = input("Enter the string: ")

# 1. Calculate the midpoint
n = len(string)
mid = n // 2

# 2. Slice into two halves
# For 'khokho', n=6, mid=3. 
# first_half = string[0:3] -> 'kho'
# second_half = string[3:6] -> 'kho'
first_half = string[:mid]
second_half = string[mid:]

# 3. Check for symmetry
if first_half == second_half:
    print("The entered string is symmetrical")
else:
    print("The entered string is NOT symmetrical")
```

```python
"""
Problem 17: Reverse words in a given String
Statement: We are given a string and we need to reverse words of a given string.

Example 1:
Input: geeks quiz practice code
Output: code practice quiz geeks
Example 2:
Input: my name is laxmi
Output: laxmi is name my
"""
# Input the string
input_str = input("Enter the string: ")

# 1. Split the string into a list of words
words = input_str.split()

# 2. Reverse the list of words
# We use slicing [::-1] which is the most efficient way to reverse a list
reversed_words = words[::-1]

# 3. Join the words back into a single string with spaces
output_str = " ".join(reversed_words)

print("Output:", output_str)
```

```python
"""
Problem 18: Find uncommon words from two Strings.
Statement: Given two sentences as strings A and B. The task is to return a list of all uncommon words.
A word is uncommon if it appears exactly once in any one of the sentences, and does not appear in the other sentence.
Note: A sentence is a string of space-separated words. Each word consists only of lowercase letters.

Example 1:
Input:
A = "apple banana mango" 
B = "banana fruits mango"
Output:
['apple', 'fruits']
"""
a = "apple banana mango"
b = "banana fruits mango"

# 1. Split both strings into words and combine them into one list
all_words = a.split() + b.split()

# 2. Use a dictionary to count the frequency of each word
word_counts = {}

for word in all_words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# 3. Filter the words that appear exactly once
uncommon_words = []
for word in word_counts:
    if word_counts[word] == 1:
        uncommon_words.append(word)

print("Output:", uncommon_words)
```

```python
"""
Problem 19: Word location in String.
Statement: Find a location of a word in a given sentence.

Example 1:
Input:
Sentence: We can learn data science through campusx mentorship program.
word: campusx
Output:
Location of the word is 7.
Note- Don't use index/find functions
"""

sentence = "We can learn data science through campusx mentorship program."
target_word = "campusx"

# 1. Split the sentence into words
# Note: We strip the '.' from the end so 'program.' becomes 'program'
words = sentence.replace(".", "").split()

# 2. Track the location manually
location = 0
found = False

for word in words:
    location += 1  # Increment position count
    if word == target_word:
        found = True
        break  # Stop the loop as soon as we find it

# 3. Output the result
if found:
    print(f"Location of the word is {location}.")
else:
    print("Word not found in the sentence.")
```
