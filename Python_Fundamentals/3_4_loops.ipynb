```python
# Write a code to print a table and the base number of the table is provide by the user:

# Take the number by the user:
number = int(input("Enter the number for the table:"))

# Provide the logic for the loop:
i = 1
while i < 11:
    print (number * i)
    i = i + 1 
```

```python
# Write a code to print a table and the number if the base number of the table is provide by the user:

# Take the number by the user and store it in a variable:
number = int(input("Enter the number for the table:"))

# Provide the logic for the loop:
i = 1
while i < 11:
    print (number,'*',i,'=',number * i)
    i = i + 1 
```

```python
# while loop with else:
i = 1 
while i < 4:
    print(i)
    i = i + 1 
else:
    print("Limit Corssed")
```

```python
# write code for a number gussing game:

# Create a random number for the gussing game
import random 
ran_number = random.randint(1,100)

# Take the user input for the random gussing number
user_input = int(input("Guess the number and enter here:"))

# Count the attempt
attempt = 1

# Build the logic for the game
while user_input != ran_number:
    if user_input < ran_number:
        print("Guess higher number")
    else:
        print("Guess lower number")

    user_input = int(input("Guess the number and enter here:"))
    attempt = attempt + 1
else:
    print("Congratulation! You gess the right number:",ran_number)
    print("Total number of attempt:",attempt)
```

```python
# sequence sum problem
# 1/1! + 2/2! + 3/3! + .... + n/n!

# Take the input from the user:
n = int(input("Enter the nth term:"))

# Setup the result and factorial value:
result = 0
fact = 1

# Implementing the logic 
for i in range(1,n+1):
    fact = fact * i
    result = result + (i/fact)

# Display the result
print("Total sum of the series is:",result)
```
