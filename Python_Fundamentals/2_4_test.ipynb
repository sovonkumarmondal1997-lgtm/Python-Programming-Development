```python
"""
Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers 
without using any special python syntax.
"""
# Take two numbers as input from the user
number_1 = int(input("Enter the first_number: "))
number_2 = int(input("Enter the second number: "))

# Showing output of the user inputs:
print("Before Swapping")
print("a:", number_1)
print("b:", number_2)

# Taking a temporary variable name 'temp_var' and swap the value of two input
temp_var = number_1
number_1 = number_2
number_2 = temp_var

# Print the final output:
print("After Swapping")
print("a:", number_1)
print("b:", number_2)
```

```python
# Q4:- Write a program to find the euclidean distance between two coordinates.
# Take both the coordinates from the user as input.

# Taking the coordinates as input from the user:
X1 = int(input("Enter the x1 value for the first_coordinate: "))
Y1 = int(input("Enter the y1 value for the first_coordinate: "))
X2 = int(input("Enter the x2 value for the second_coordinate: "))
Y2 = int(input("Enter the y2 value for the second coordinate: "))

# Finding the eucildian distance between two coordinates:
euclidian_distance = ((X2 - X1)**2 + (Y2 - Y1)**2)**0.5

# Print the final output:
print("Euclidian distance between two coordinates is: ",round(euclidian_distance,2))
```

```python
# Q5:- Write a program to find the simple interest when the value of principle,
# rate of interest and time period is provided by the user.

# Take the input from the user:
P = int(input("Enter the principal value: "))
R = float(input("Enter the rate of interest (%): "))
T = int(input("Enter the time (in years): "))

# Calculate the simple interest:
simple_interest = (P * T * R) / 100

# Calculate the total amount:
total_amount = P + simple_interest

# Display the result:
print("The simple interest is: ",simple_interest)
print("The total amount is: ",total_amount)
```

```python
# Q7:- Write a program to find the sum of squares of first n natural numbers,
# where n will be provided by the user.

# Take the input from the user:
n = int(input("Enter the value of n: "))

# Calculate the sum of squares:
sum_of_squares = (n * (n + 1) * (2* n + 1)) / 6

# Display the result:
print("Sum of squares of first n naural number is: ",sum_of_squares)
```

```python
# Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series,
# Assume all inputs are provided by the user.

# Take the output:
a = int(input("Enter the first term of the arithmetic series (a): "))
b = int(input("Enter the second term of the arithmetic series (b): "))
n = int(input("Enter the no of terms of the series (n): "))

# calculate the difference between first and second term:
d = b - a 

# Calculate the n_th term of the series:
n_th_term = a + (n - 1) * d

# Display the result:
print("The n_th term of the series is: ",n_th_term)
```

```python
# Q9:- Given 2 fractions, find the sum of those 2 fractions.
# Take the numerator and denominator values of the fractions from the user.

# Take the input from the user:
a = int(input("Enter the numarator value (a) for first_fraction: "))
b = int(input("Enter the denominator value (b) for the first_fraction: "))
c = int(input("Enter the numarator value (c) for the second_fraction: "))
d = int(input("Enter the denominator value (d) for the second_fraction: "))

# Calculate the numarator and denominator of two fraction:
nm = (a * d + b * c)
dm = b * d

# Display the result:
print("The fraction value is:{0}/{1}".format(nm,dm))
```

```python
"""
Q10:- Given the height, width and breadth of a milk tank,
you have to find out how many glasses of milk can be obtained?
Assume all the inputs are provided by the user.

Input:
Dimensions of the milk tank
H = 20cm, L = 20cm, B = 20cm

Dimensions of the glass
h = 3cm, r = 1cm
"""

# Take the input from the user:
H = int(input("Enter the value of hight (H) for the milk tank: "))
L = int(input("Enter the value of length (L) for the milk tank: "))
B = int(input("Enter the value of breadth (B) for the milk tank: "))
h = int(input("Enter the value of hight (h) for the glass: "))
r = int(input("Enter the value of radius (r) for the glass: "))

# Calculate the volume of the milk tank:
volume_milk_tank = H * L * B

# Calculate the volume of the glass:
volume_glass = 3.14 * (r*r) * h

# Calculate the no of glasses:
no_of_glasses = (volume_milk_tank // volume_glass)

# Display the result:
print("No of glasses milk obtained: ",no_of_glasses)
```
